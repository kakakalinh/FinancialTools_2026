### Introduction to Derivatives

*   A derivative is a financial instrument whose payoffs and values are derived from or depend on an underlying asset.
    *   Futures and Forwards
    *   Options
    *   Swaps

*   The underlying asset can be:
    *   Stocks & Stock indices
    *   Bonds
    *   Energy & Commodities
    *   Currencies

*   Derivatives play a key role in transferring risks in the economy.
*   A derivative may be,
    *   A standalone instrument
    *   Embedded in other products

---

### What are derivatives used for?

*   **Hedging**
    *   Risk management: To cut away unwanted portions of risk exposure.

*   **Speculation**
    *   Used to bet on the future direction of an asset's price, often with no offsetting exposure.

*   To lock in an arbitrage profit.
*   To change the nature of a liability or an investment.

---

### Types of Markets

*   **Exchange Traded**
    *   Standardized contracts defined by an exchange.
    *   Examples: Chicago Mercantile Exchange (CME), Chicago Board Options Exchange (CBOE).

*   **Over-the-counter (OTC)**
    *   Telephone and computer-linked network of dealers/brokers.
    *   More flexibility, but introduces credit/counterparty risk.

### Forward Contracts

*   A forward contract is an agreement (contract) to buy or sell an asset at a certain future time for a certain price
    *   An agreement (Forward contract)
    
    *   To buy / sell an asset (trade Underlying asset)
    
    *   At future time (Maturity)
    
    *   For a certain price (Forward price)

*   The buyer and the seller are obligated to perform under the terms of the contract.

*   OTC Transactions

---


### Forward Contracts

*   Two parties agree to exchange some item in the future at a delivery price specified now.

*   The forward price is defined as the delivery price that makes the current market value of the contract zero.
    *   No money is paid in the present by either party to the other.

*   The face value of the contract is the quantity of the item specified in the contract times the forward price.

*   The party who agrees to buy the specified item is said to take a long position, and the party who agrees to sell the item is said to take a short position.

---

### Forward Contract Payoffs

*   For $S_T$ = spot price at maturity, and F = Forward price;
    
    *   The payoff from a long position in a forward contract is: $S_T - F$
    
    *   The payoff from a short position in a forward contract is: $F – S_T$

### Forward Contracts & Replication

*   **Portfolio A:** Buy one share of the stock.
*   **Portfolio B (Synthetic Stock):** Go long one forward contract + Buy a risk-free, zero-coupon bond with a face value of F.

| Position                                 | Immediate Cash Flow | Cash Flow at Expiration (Time T) |
| :--------------------------------------- | :-----------------: | :------------------------------: |
| **Portfolio A: Stock**                   |        $-S_0$    |               $S_T$              |
| **Portfolio B: Synthetic Stock**         |                     |                                  |
| 1. Long a forward contract               |          0          |            $S_T - F$             |
| 2. Buy a bond with face value F          |   $-F / (1+r)^T$    |               +$F$                |
| **Total for Portfolio B**                | $-F / (1+r)^T$  |               $S_T$             |

Since both portfolios have the same payoff ($S_T$) at time T, they must have the same cost today.

$ \text{Cost of Portfolio A} = \text{Cost of Portfolio B} $

$ S_0 = \frac{F}{(1+r)^T} \quad \Rightarrow \quad F = S_0(1 + r)^T $

**Conclusion:**
Rearranging the formula gives us the no-arbitrage forward price:
$$ F = S_0(1 + r)^T $$

The forward price equals the future value of the spot price compounded at the risk-free interest rate for T years.

### Forward Contracts

A trader has a short forward contract on 100 million JPY at a forward price (F) of $0.0080/JPY.

*   Payoff Formula: $(F – S_T)$ × Contract Size

*   If the spot rate $(S_T)$ at expiration is \$0.0074/JPY: $\quad \Rightarrow \quad \text{Payoff} = (\$0.0080 – \$0.0074) × 100M = +\$60,000$ (Profit)

*   If the spot rate $(S_T)$ at expiration is \$0.0091/JPY: $\quad \Rightarrow \quad \text{Payoff} = (\$0.0080 – \$0.0091) × 100M = -\$110,000$ (Loss)

**Arbitrage Identification**

The price of gold is $1,800/oz. A one-year forward contract is trading at $2,000/oz. The risk-free rate is 5%.

*   Calculate the fair (no-arbitrage) forward price: $\quad \Rightarrow \quad F = S_0(1 + r) = \$1,800 × 1.05 = \$1,890$

*   Compare the market price to the fair price. The market forward price $(\$2,000)$ is higher than the fair price $(\$1,890)$. The contract is overpriced.

*   Formulate the arbitrage strategy.
    *   The strategy is to sell the overpriced item (the forward contract) and simultaneously buy the underlying asset with borrowed funds.

| Action                          | Cash Flow Today (T=0) | Cash Flow in One Year (T=1)            |
| :------------------------------ | :--------------------: | -------------------------------------: |
| 1. Borrow funds                 | +$1,800               | -$1,800 × 1.05 = -$1,890               |
| 2. Buy gold in spot market      | -$1,800               | (Hold one ounce of gold)               |
| 3. Sell a forward contract      | $0                    | Sell gold, receive +$2,000             |
| **Net Cash Flow**               | **$0**                | **-$1,890 + $2,000 = +$110**            |

This transaction requires no initial investment and generates a guaranteed, risk-free profit of $110 in one year.

### Futures Contracts

*   A futures contract is a standardized forward contract traded on an exchange.

*   Specifications are defined by the exchange:

    *   What can be delivered,

    *   Where it can be delivered,

    *   When it can be delivered.

*   Futures contracts are settled daily (Marked-to-Market).
---

### Futures Contracts: Margin

*   To minimize default risk, the exchange requires a margin deposit from both parties.

*   A margin is cash or marketable securities deposited by an investor with his or her broker.
    *   The balance in the margin account is adjusted to reflect daily settlement.
    *   Margins minimize the possibility of a loss through a default on a contract.

*   Upon contract initiation, both parties open margin accounts.

*   Associated with each type of contract is an initial margin and a maintenance margin.

*   The initial deposit of funds into the margin account is known as initial margin.

*   The maintenance margin is less than the initial margin.

*   If the balance in your margin account falls below the maintenance margin, you will receive a margin call that requires you to increase your margin account back to the initial margin level.

### Introduction to Options

*   Options are an example of a broader class of assets called contingent claims.
    
    *   A contingent claim is any asset whose future payoff is contingent on the outcome of some uncertain event.

*   A financial option contract gives its owner the right (but not the obligation) to purchase or sell an asset at a fixed price at some future date.
    
    *   Call options
    
    *   Put options
    
    *   Option buyer (holder), Option seller (writer)
    
    *   Long position (right), short position (obligation)
    
    *   Option premium: The market price of the option.
    
        *   Compensates the seller for the risk of loss in the event that the option holder chooses to exercise the option.

---

### Option Terminology

*   Strike Price (or Exercise Price, K): The fixed price at which the holder can buy or sell.

*   Expiration Date: The date the option expires.

*   Moneyness:
    *   In-the-money: Payoff from immediate exercise is positive.
        *   Call: $S > K$
        *   Put: $K > S$
    
    *   Out-of-the-money: Payoff from immediate exercise is negative.
    
    *   At-the-money: Exercise price equals the asset price ($S=K$).

---

### Option Payoffs at Expiration (T)

*   The analytical payoff for a **Long Call** is: $\qquad Payoff = max(S_T – K, 0)$

*   The analytical payoff for a **Long Put** is: $\qquad Payoff = max(K – S_T, 0)$

### A Simple Strategy: The Protective Put

A protective put consists of owning the underlying asset and buying a put option to protect against a price decline.

|              | Payoff if $S_T \le K$ | Payoff if $S_T > K$ |
| :----------- | :----------: | :--------: |
| **Stock**    | $S_T$       | $S_T$     |
| **+Put**     | $K-S_T$     | 0         |
| **Total**    | **K**       | **$S_T$**  |

This strategy creates an "insurance" payoff floor at the strike price, K.

### Put-Call Parity

The Law of One Price states that two portfolios with the same payoffs must have the same price.

*   Portfolio 1: Stock + Put Option

*   Portfolio 2: Risk-Free Bond + Call Option

Both portfolios have the same payoff at expiration: $max(S_T, K)$. Therefore, their prices today must be equal.

### Put-Call Parity Formula

This relationship gives us the Put-Call Parity equation for European options:

$S_0 + P = C + PV(K)$

*   Calls are equivalent to: long put, long stock, and borrowing the PV of the strike price.
    $C = S_0 + P - PV(K)$

*   Puts are equivalent to: long call, short stock, and lending the PV of the strike price.
    $P = C - S_0 + PV(K)$

### Put-Call Parity

*   You are an options dealer specializing in bespoke derivative contracts. 
    *   A client wants to purchase a one-year European call option on Sigma Corp stock with a strike price of $\$65$.
    
    *   Another dealer is offering to sell you a one-year European put option on Sigma Corp stock with the same strike price of $\$65$ for a price of $\$9.00$ per share. 
    
    
    *   Sigma Corp pays no dividends and is currently trading at $\$60$ per share. If the risk-free interest rate is 4%, what is the minimum price you must charge for the call option to avoid a loss and guarantee a profit at any higher price?

    $$C = S_0 + P - \frac{K}{(1 + r_f)^T} = 60 + 9.00 - \frac{65}{(1 + 0.04)^1} = \$6.50$$

**Replication and Cash Flow Table:**

| Action | Cash Flow at $t=0$ | Payoff if $S_1 \le \$65$ | Payoff if $S_1 > \$65$ |
| :----------------- | :-------------: | :-------------: | :-------------: |
| Buy Put Option     | $-\$9.00$       | $65 - S_1$      | $0$             |
| Buy Stock          | $-\$60.00$      | $S_1$           | $S_1$           |
| Borrow PV of Strike (Sell Bond) | $+\$62.50$ | $-\$65.00$      | $-\$65.00$      |
| **Synthetic Call Portfolio** | **$-\$6.50$** | **$0$** | **$S_1 - 65$** |
| Sell Call Option (to Client) | $+\$6.50$ | $0$ | $-(S_1 - 65)$ |
| **Total Net Cash Flow** | **$0$** | **$0$** | **$0$** |

### Binomial Option Pricing Model

A simple but powerful model for pricing options based on the no-arbitrage principle.

*   In a single time period, the stock price can only move to one of two possible states: an "Up" state or a "Down" state.

*   We create a replicating portfolio of the underlying stock and risk-free borrowing/lending that has the exact same payoffs as the option in both states.

*   The price of the option today must equal the cost of this replicating portfolio.

---


### Binomial Option Pricing Model: Example

* A 3-month call option on a stock has a strike price of **21**. Risk-free rate is 12%.

    * **Current State:** Stock price = $20, Option Price = ?
    
    * **State after 3 months (Up):** Stock Price = $22, Option Price = $max(22 - 21, 0) = $1
    
    * **State after 3 months (Down):** Stock Price = $18, Option Price = $max(18 - 21, 0) = $0

### Binomial Model: The Riskless Portfolio

* Consider the portfolio: long $\Delta$ shares, short 1 call option
    *   Value in Up state: $22\Delta - 1$
    
    *   Value in Down state: $18\Delta - 0$

* The portfolio is riskless when its value is the same in both states:
    
    $22\Delta - 1 = 18\Delta$
    
    $4\Delta = 1$
    
    $\Delta = 0.25$

This $\Delta$ is the hedge ratio.

---

### Binomial Model: Pricing the Option

* The riskless portfolio is: long 0.25 shares, short 1 call option.

* The value of this portfolio in 3 months is certain: $18 \times 0.25 = \$4.50$.

* The value of this portfolio today must be the present value of $4.50:
    *   Value Today = $4.50e^{-0.12 \times 0.25} = \$4.367$

* The value of the shares held in the portfolio today is: $0.25 \times \$20 = \$5.00$.

* The portfolio value today is also (Value of Shares - Value of Option):
    *   $4.367 = 5.00 - \text{Call Price (C)}$

* Therefore, $\quad C = 5.00 - 4.367 = \$0.633$
  
---

### Binomial Model: Risk-Neutral Valuation

An alternative approach is to calculate "risk-neutral probabilities".

*   Let $p_u$ be the risk-neutral probability of an up move and $p_d = (1-p_u)$ be the probability of a down move.

    $$p_u = \frac{e^{r\Delta t} - d}{u - d}$$

*   The option price is then the discounted expected payoff using these probabilities.

    $$f_0 = e^{-r\Delta t}[p_u f_u + (1-p_u)f_d]$$

*   This method gives the exact same price as the replication method.
  
---

### Binomial Option Pricing Model

* Consider a European call option that expires in one period with an exercise price of $105. Assume the stock price today ($S_0$) is $100. In one period, the stock price will either rise by $20 or fall by $20. The one-period risk-free rate is 5%.

    * $S_0 = 100, f_0 = ?$
    * Up state: $S_u = 120, f_u = 15$ (since $120 - 105 = 15$). Factor $u = 120/100 = 1.20$.
    * Down state: $S_d = 80, f_d = 0$ (since $80 < 105$). Factor $d = 80/100 = 0.80$.
    * Risk-free gross return $r = 1.05$.
    * Replicating portfolio consisting of $\Delta$ shares and $B$ in risk-free bonds:
        $$ f_u = 15 = \Delta S_0 u + Br = \Delta \cdot 120 + B \cdot 1.05 $$
        $$ f_d = 0 = \Delta S_0 d + Br = \Delta \cdot 80 + B \cdot 1.05 $$

    $$ \Delta = \frac{f_u - f_d}{S_0(u - d)} = \frac{15 - 0}{100(1.20 - 0.80)} = \frac{15}{40} = 0.375 $$

  
    $$ B = \frac{u f_d - d f_u}{r(u - d)} = \frac{1.20 \cdot 0 - 0.80 \cdot 15}{1.05(1.20 - 0.80)} = \frac{-12}{0.42} \approx -28.571 $$

    **The replicating portfolio consists of buying 0.375 shares and borrowing $28.571 at the risk-free rate.**

    * **Option Value Today:**
        $$ \text{Cost} = \Delta S_0 + B = 0.375 \cdot 100 + (-28.571) = 37.50 - 28.571 = \$8.929 $$

### Binomial Option Pricing Model

* Consider a European call option that expires in one period with an exercise price of $110. Assume that the current stock price ($S_0$) is $100. In one period, the stock price will either rise by $30 or fall by $10. The one-period risk-free rate is 8%.

    * $S_0 = 100, f_0 = ?$
    
    * Up state: $S_u = 130, f_u = 20$. ($u = 130/100 = 1.30$)
    
    * Down state: $S_d = 90, f_d = 0$. ($d = 90/100 = 0.90$)
    
    * Risk-free gross return $r = 1.08$

    * Using risk-neutral probabilities:
        $$ p_u = \frac{r - d}{u - d} = \frac{1.08 - 0.90}{1.30 - 0.90} = \frac{0.18}{0.40} = 0.45 $$
        $$ p_d = 1 - p_u = 0.55 $$
        $$ f_0 = \frac{p_u f_u + p_d f_d}{r} = \frac{0.45 \times 20 + 0.55 \times 0}{1.08} = \frac{9}{1.08} \approx \$8.333 $$
