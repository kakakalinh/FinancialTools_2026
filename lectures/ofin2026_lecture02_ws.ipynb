### The Time Value of Money

* $1,000 is deposited into a deposit account that pays 12% annual interest.

  * Value after 1 year? Value after 2 years? Value after 3 years?

    $$FV_1 = 1,000.00 \times (1 + 0.12) = 1,000.00 \times 1.12 = \$1,120.00$$

    $$FV_2 = 1,120.00 \times (1 + 0.12) = 1,120.00 \times 1.12 = \$1,254.40$$

    $$FV_3 = 1,254.40 \times (1 + 0.12) = 1,254.40 \times 1.12 = \$1,404.93$$

|          | Year 1   | Year 2   | Year 3   | Total    |
|----------|----------:|----------:|----------:|----------:|
| Principal| \$1,000.00| \$1,000.00| \$1,000.00|          |
| Interest | \$120.00  | \$120.00  | \$120.00  | \$360.00  |
| Compound |   | \$14.40   | \$14.40   | \$28.80   |
| Extra    |          |    | \$1.73    | \$1.73    |
| **Total**|          |          |          | **\$404.93** |

* **Total Interest (Compounded):** $\qquad \boxed{\$404.93}$
* **Simple Interest:** $\qquad \qquad  \qquad \qquad 1,000.00 \times (0.12 \times 3) = \boxed{\$360.00}$
* **Power of Compounding:** $\qquad \qquad \boxed{\$44.93}$

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import base64
import io
from IPython.display import display, HTML, clear_output
import ipywidgets as widgets
from IPython.display import Javascript

# Define the function to calculate simple interest FV
def simple_interest_fv(PV, r, t):
    return PV * (1 + r * t)

# Define the function to calculate compound interest FV
def compound_interest_fv(PV, r, t, m):
    return PV * (1 + r / m) ** (m * t)

# Function to generate and display the output
def generate_output(PV, t, r, m1, m2, output_widget):
    with output_widget:
        try:
            # Clear the output area for the plots and tables
            clear_output(wait=True)

            # Time periods for plotting
            time_periods = np.arange(1, t + 1)

            # Calculate FV for investment (constant)
            investment = np.full_like(time_periods, PV, dtype=float)

            # Calculate FV for simple and compound interest
            simple_fv = simple_interest_fv(PV, r, time_periods)
            compound_fv_m1 = compound_interest_fv(PV, r, time_periods, m1)
            compound_fv_m2 = compound_interest_fv(PV, r, time_periods, m2)

            # Final FV values for the tables (at t)
            final_compound_fv_m1 = compound_interest_fv(PV, r, t, m1)
            final_compound_fv_m2 = compound_interest_fv(PV, r, t, m2)

            # Create a single figure with two subplots side by side
            fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5), sharey=True)

            # Plot 1: m=m1
            ax1.fill_between(time_periods, investment, simple_fv, color='yellow', label='Simple', alpha=0.7)
            ax1.fill_between(time_periods, simple_fv, compound_fv_m1, color='gray', label='Compound', alpha=0.7)
            ax1.fill_between(time_periods, 0, investment, color='blue', label='Investment', alpha=0.7)
            ax1.set_title(f"(t={t}, m={m1}, r={int(r*100)}%)")
            ax1.set_xlabel("t")
            ax1.set_ylabel("FV ($)")
            ax1.legend(loc="upper left")
            ax1.grid(True)

            # Plot 2: m=m2
            ax2.fill_between(time_periods, investment, simple_fv, color='yellow', label='Simple', alpha=0.7)
            ax2.fill_between(time_periods, simple_fv, compound_fv_m2, color='gray', label='Compound', alpha=0.7)
            ax2.fill_between(time_periods, 0, investment, color='blue', label='Investment', alpha=0.7)
            ax2.set_title(f"(t={t}, m={m2}, r={int(r*100)}%)")
            ax2.set_xlabel("t")
            ax2.legend(loc="upper left")
            ax2.grid(True)

            # Adjust layout to prevent overlap
            plt.tight_layout()

            # Save the figure to a buffer and close it
            buf = io.BytesIO()
            plt.savefig(buf, format='png', bbox_inches='tight')
            buf.seek(0)
            image_base64 = base64.b64encode(buf.getvalue()).decode('utf-8')
            plt.close(fig)
            buf.close()

            # Create image HTML
            image_html = f'<img src="data:image/png;base64,{image_base64}" style="width: 90%;"/>'

            # Create summary tables
            table_data_m1 = {
                "PV": [f"{PV:.2f}"],
                "t": [t],
                "m": [m1],
                "r": [f"{int(r*100)}%"],
                "FV": [f"{final_compound_fv_m1:.2f}"]
            }
            table_data_m2 = {
                "PV": [f"{PV:.2f}"],
                "t": [t],
                "m": [m2],
                "r": [f"{int(r*100)}%"],
                "FV": [f"{final_compound_fv_m2:.2f}"]
            }

            df_m1 = pd.DataFrame(table_data_m1)
            df_m2 = pd.DataFrame(table_data_m2)

            table_html_m1 = df_m1.to_html(index=False, float_format=lambda x: f"{x:.2f}")
            table_html_m2 = df_m2.to_html(index=False, float_format=lambda x: f"{x:.2f}")

            # Combine image and tables in HTML
            html_content = f"""
            <div style="display: flex; justify-content: space-between; align-items: flex-start; width: 100%;">
                <div style="flex: 1; text-align: center;">
                    <div>{image_html}</div>
                    <div style="display: flex; justify-content: space-around;">
                        <div style="flex: 1; margin-right: 10px;">{table_html_m1}</div>
                        <div style="flex: 1;">{table_html_m2}</div>
                    </div>
                </div>
            </div>
            """

            # Display new content in the output widget
            display(HTML(html_content))

        except Exception as e:
            clear_output(wait=True)
            display(HTML(f"<p style='color: red;'>Error: {str(e)}</p>"))

# Create interactive widgets
PV_widget = widgets.FloatText(value=1000, description='PV ($):', step=100)
t_widget = widgets.IntText(value=25, description='t (years):', min=1, max=50)
r_widget = widgets.FloatText(value=10, description='r (%):', step=0.1)
m1_widget = widgets.IntText(value=1, description='m1:', min=1, max=12)
m2_widget = widgets.IntText(value=2, description='m2:', min=1, max=12)

# Create regenerate button
button = widgets.Button(description="Regenerate")

# Create an output widget for the plots and tables
output_widget = widgets.Output()

# Define button click event
def on_button_clicked(b):
    PV = PV_widget.value
    t = t_widget.value
    r = r_widget.value / 100  # Convert percentage to decimal
    m1 = m1_widget.value
    m2 = m2_widget.value
    generate_output(PV, t, r, m1, m2, output_widget)

button.on_click(on_button_clicked)

# Organize widgets in a box
widget_box = widgets.VBox([
    PV_widget,
    t_widget,
    r_widget,
    m1_widget,
    m2_widget,
    button
])

# Display the widgets and output separately
display(widget_box)
display(output_widget)

# Initial generation
generate_output(1000, 25, 0.10, 1, 2, output_widget)
```

### The Time Value of Money

* The future value of the $1,000 deposit in an account paying 10% interest annually at the end of 10 years:

    $FV = PV (1 + i)^n \qquad FV = 1,000 (1 + 0.10)^{10} \qquad FV = \$2,593.74$

* What is the annual return on this deposit?

    The total return over 10 years:
    $$
    \frac{2,593.74 - 1,000}{1,000} = 1.59374
    $$
    Average return per year:
    $$
    \frac{1.59374}{10} = 0.159374 \, \text{or} \, 15.94\%
    $$

    The exact annual return (geometric mean):
    $$
    \sqrt[10]{\frac{2,593.74}{1,000}} - 1 = \sqrt[10]{2.59374} - 1 = 0.10 \, \text{or} \, 10.00\%
    $$

### Valuing a stream of cash flows
   * When the payments are made (beginning vs. end of the period) is important.
   * Standard formulas apply to regular payments made at regular intervals.
   * Future value vs. present value formulas.

**Future value of a fixed amount paid at the end of each period**
$$
FV = CF \left[ \frac{(1+i)^n - 1}{i} \right]
$$

**Future value of a fixed amount paid at the beginning of each period**
$$
FV = CF \left[ \frac{(1+i)^n - 1}{i} \right] \times (1 + i)
$$

**Present value of a fixed amount paid at the end of each period**
$$
PV = CF \left[ \frac{1 - \frac{1}{(1+i)^n}}{i} \right]
$$

**Present value of a fixed amount paid at the beginning of each period**

$\qquad \qquad \qquad \qquad \qquad \qquad \qquad PV = CF [\frac{1 - \frac{1}{(1+i)^{n-1}}}{i}] + \text{CF} \qquad \text{or} \qquad PV = CF [ \frac{1 - \frac{1}{(1 + i)^n}}{i}] \times (1 + i)$

* Suppose that you take out a five-year loan of $\$15,000$. 
    * The bank requires you to repay the loan evenly (in equal annual installments) over the five years. 
    * If the interest rate is 8\% a year, what should the regular annual payments be?
* Now, suppose that you decide to make a one-time $\$4,000$ balloon payment at the end of the 3rd year.  
    * What should your newly adjusted regular annual payments be to fully pay off the loan?

---

### Exercise 2.1: Solution

**Calculate the standard regular annual payments**

$$ PV = CF \times \frac{1 - \frac{1}{(1 + r)^n}}{r} $$

Loan details: $PV = \$15,000$, $r = 0.08$, $n = 5$
$$ 15000 = CF \times \frac{1 - \frac{1}{(1.08)^5}}{0.08} \quad \rightarrow \quad 15000 = CF \times \frac{1 - 0.680583}{0.08} \quad \rightarrow \quad 15000 = CF \times 3.992712 \quad \rightarrow \quad CF = \$3,756.85 $$

---

**Calculate the new annual payments with the balloon payment**

Find the present value of that single balloon payment today (Year 0):
$$ PV_{\text{balloon}} = \frac{FV}{(1 + r)^t} = \frac{\$4,000}{(1.08)^3} = \frac{\$4,000}{1.259712} = \$3,175.33 $$

Subtract the present value of the balloon payment from the original loan amount. This gives us the remaining loan amount that must be covered by the regular annual payments:
$$ PV_{\text{remaining}} = \$15,000 - \$3,175.33 = \$11,824.67 $$

Calculate the new annual payments (CF) needed to pay off this remaining present value over the 5 years:
$$ 11824.67 = CF \times \frac{1 - \frac{1}{(1.08)^5}}{0.08} \quad \rightarrow \quad 11824.67 = CF \times 3.992712 \quad \rightarrow \quad  CF = \frac{11824.67}{3.992712} = \$2,961.56 $$

With the balloon payment, the newly adjusted regular annual payments should be **$\$2,961.56$**.

### The Time Value of Money: Perpetuities

* When a constant cash flow will occur at regular intervals forever it is called a perpetuity.

* The value of a perpetuity is simply the cash flow divided by the interest rate.
$$
PV (C \ in \ Perpetuity) = \frac{C}{i}
$$

**Example 2.2**
  * You are planning to establish a scholarship fund for underrepresented students in engineering at your university. The fund should provide \$120,000 per year to support scholarships, research grants, and conference travel expenses. If the fund is expected to generate an annual return of 6%, how much money do you need to donate initially to sustain this yearly expenditure indefinitely?

$$PV = \frac{C}{r}$$
where:
- $ PV $: Present Value (initial donation needed)
- $ C $: Annual cash flow (\$120,000)
- $ r $: Annual rate of return (6% or 0.06)

$\qquad PV = \frac{120,000}{0.06} \qquad \rightarrow \qquad PV = 2,000,000 \qquad \rightarrow \qquad \text{The initial donation required is \$2,000,000}$

---

```python
# Annual cash flow required for the scholarship fund
C = 120000  

# Annual rate of return (as a decimal)
r = 0.06  

# Calculate the present value using the perpetuity formula
PV = C / r

print("The initial donation required is: ${:,.0f}".format(PV))
```

### The Time Value of Money: Growing Perpetuity
* Assume you expect the amount of your perpetual payment to increase at a constant rate, $g$. Then,  
$$PV (C \ \text{in} \ \text{Perpetuity}) = \frac{C}{i - g}$$

* In the previous example, the university has asked you to increase the donation to account for the effect of inflation on scholarships and grants, which is expected to be 2.5% per year. How much more will you need to donate to satisfy that request?  
$$PV (C \ \text{in} \ \text{Perpetuity}) = \frac{\$120,000}{0.06 - 0.025} = \frac{\$120,000}{0.035} = \$3,428,571.43$$

---

```python
# Annual cash flow required for the scholarship fund
C = 120000  

# Nominal annual rate of return (as a decimal)
r = 0.06  

# Expected annual inflation (growth) rate (as a decimal)
g = 0.025  

# Calculate the present value required for a growing perpetuity
PV_growing = C / (r - g)

# Calculate the nominal (non-growing) perpetuity required
PV_nominal = C / r

# Calculate the additional donation needed to fund a growing perpetuity
additional_donation = PV_growing - PV_nominal

print("Present value of growing perpetuity: ${:,.0f}".format(PV_growing))
print("Present value of perpetuity (without growth): ${:,.0f}".format(PV_nominal))
print("Additional donation needed: ${:,.0f}".format(additional_donation))
```

### The Time Value of Money: Growing Annuity

*   The future value of a growing annuity with the initial cash flow, $C$, growth rate, $g$, and interest rate, $i$, is defined as:
$$
FV = C \times \frac{(1 + i)^N - (1 + g)^N}{i - g}
$$
*   The present value of a growing annuity is defined as:
$$
PV = \frac{C}{i - g} \times \left[ 1 - \left( \frac{1 + g}{1 + i} \right)^N \right]
$$

**Example** 
* You want to start saving for a down payment on a house. You plan to contribute $\$5,000$ to an investment account at the end of this year. You anticipate that you will be able to increase your annual contributions by 5% each year for the next 10 years. If your investment account is expected to generate an annual return of 7%, how much will you have saved for your down payment in 10 years?

$$
FV = 5,000 \times \frac{(1.07)^{10} - (1.05)^{10}}{0.07 - 0.05} = 5,000 \times \frac{1.96715 - 1.62889}{0.02} \approx \$84,564
$$

* What is the value of this savings plan in today's dollars (its Present Value)?

$$
PV = \frac{5,000}{0.07 - 0.05} \times \left[ 1 - \left( \frac{1.05}{1.07} \right)^{10} \right] = 250,000 \times 0.171953 \approx \$42,988
$$

---

```python
# Parameters for the growing annuity
C = 5000    # Initial contribution at the end of the first year
r = 0.07     # Expected annual return (8%)
g = 0.05     # Annual growth rate of contributions (4%)
n = 10       # Number of years of contributions

# Calculate the future value of the growing annuity using the formula:
# FV = C * [ (1 + r)^n - (1 + g)^n ] / (r - g)
FV = C * ((1 + r)**n - (1 + g)**n) / (r - g)

print("Future Value of your retirement savings: ${:,.2f}".format(FV))
```

```python
import ipywidgets as widgets
from ipywidgets import interact
import numpy as np
import matplotlib.pyplot as plt

def calculate_future_value(C, r, g, n):
    """
    Calculate the future value of retirement savings using a growing annuity and plot the savings growth over time.
    
    Parameters:
    C : float
        Initial contribution at the end of the first year.
    r : float
        Expected annual return (as a decimal).
    g : float
        Annual growth rate of contributions (as a decimal).
    n : int
        Number of years of contributions.
    """
    # Calculate the future value of the growing annuity
    FV = C * ((1 + r)**n - (1 + g)**n) / (r - g)
    
    # Calculate the savings growth over time
    years = np.arange(n + 1)  # Time points from 0 to n
    future_values = np.zeros(n + 1)  # Array to store future values at each year
    
    for t in range(1, n + 1):
        future_values[t] = C * ((1 + r)**t - (1 + g)**t) / (r - g)
    
    # Print the future value at retirement
    print("Future Value of your retirement savings: ${:,.2f}".format(FV))
    
    # Plot the savings growth over time
    plt.figure(figsize=(8, 5))
    plt.plot(years, future_values, marker='o', linestyle='-', color='b', label="Future Value Over Time")
    plt.xlabel("Years")
    plt.ylabel("Future Value ($)")
    plt.title("Growth of Retirement Savings Over Time")
    plt.grid(True)
    plt.legend()
    plt.show()

# Create interactive widgets for input parameters
interact(
    calculate_future_value,
    C=widgets.FloatText(value=5000, description='Initial Contribution ($):'),
    r=widgets.FloatSlider(value=0.07, min=0.0, max=0.2, step=0.01, description='Annual Return (r):'),
    g=widgets.FloatSlider(value=0.05, min=0.0, max=0.2, step=0.01, description='Growth Rate (g):'),
    n=widgets.IntSlider(value=10, min=1, max=100, step=1, description='Years (n):')
)
```

### Exercise 2.2

In September 2019, a Californian man won a record $\$680$ million lottery prize, to be paid out over 30 annual installments. 

The first payment was \$10.235 million, and each subsequent payment increased by 5% per year, so that the final payment is approximately \$42.13 million. 

If the annual interest rate is 3%, what is the present value of this $\$680$ million prize?


---

### Exercise 2.2: Solution

Let $P_0 = 10,235,000$ be the first payment at $t=0$.

Let $g = 0.05$ be the annual growth rate.

Let $r = 0.03$ be the annual interest rate.

Let $n = 30$ be the number of annual payments at $t=0, 1, \dots, 29$.

The present value (PV) of a growing annuity due is given by:
$$PV = P_0 \sum_{k=0}^{n-1} \left(\frac{1+g}{1+r}\right)^k = P_0 \frac{1 - \left(\frac{1+g}{1+r}\right)^n}{1 - \frac{1+g}{1+r}}$$
Let $c = \frac{1+g}{1+r} = \frac{1.05}{1.03}$.
$$PV = 10,235,000 \frac{1 - (\frac{1.05}{1.03})^{30}}{1 - \frac{1.05}{1.03}}, \; \text{and} \; (\frac{1.05}{1.03})^{30} \approx 1.7813, \; \text{and} \; 1 - \frac{1.05}{1.03} = \frac{1.03 - 1.05}{1.03} = \frac{-0.02}{1.03}$$
$$PV = 10,235,000 \frac{1 - 1.7813}{\frac{-0.02}{1.03}} = 10,235,000 \frac{-0.7813}{\frac{-0.02}{1.03}} \quad \Rightarrow \quad PV = 10,235,000 \times \frac{1.03}{0.02} \times 0.7813 \quad \Rightarrow \quad \approx 411,817,084.37649$$

$n = 20$ 

The payments form a growing ordinary annuity with payments at $t=1, 2, \dots, 20$.

The present value (PV) of a growing ordinary annuity is given by:

$$PV = C_1 \frac{1 - \left(\frac{1+g}{1+r}\right)^n}{r-g} \quad \Rightarrow \quad PV = 250,000 \frac{1 - \left(\frac{1.03}{1.07}\right)^{20}}{0.07-0.03} \quad \Rightarrow \quad PV \approx \$3,332,915.85$$

```python
# Parameters for the growing annuity
C = 250000    # Initial contribution at the end of the first year
r = 0.07     # Expected annual return (8%)
g = 0.03     # Annual growth rate of contributions (4%)
n = 20       # Number of years of contributions

# Calculate the future value of the growing annuity using the formula:
# FV = C * [ (1 + r)^n - (1 + g)^n ] / (r - g)
PV = C * (((1-((1+g)/(1 + r))**n) / (r - g)))

print("Present Value of your retirement savings: ${:,.2f}".format(PV))
```

### Exercise 2.4: Solution

First, we determine the target amount needed in the account at the moment Mia starts college (Year 18). Second, we calculate the required annual deposits to reach that target.

**The target amount (Future Value needed)**

$$\text{PVAD} = \text{PMT} \times \left[ \frac{1 - (1+r)^{-n}}{r} \right] \times (1+r)$$
Where:
* **PMT** (Annual Tuition): $45,000
* **r** (Interest Rate): 7% or 0.07
* **n** (Number of Years): 4

$$\text{PVAD} = \$45,000 \times \left[ \frac{1 - (1.07)^{-4}}{0.07} \right] \times (1.07) \quad \Rightarrow \quad \text{PVAD} \approx \$163,094.22$$

**Calculate fixed annual deposits**
$$\text{FV} = \text{PMT} \times \left[ \frac{(1+r)^n - 1}{r} \right]$$
Where:
* **FV** (Target Goal): $163,094.22
* **r** (Interest Rate): 7% or 0.07
* **n** (Number of Years): 18

$$\$163,094.22 = \text{PMT} \times \left[ \frac{(1.07)^{18} - 1}{0.07} \right]$$
$$\$163,094.22 = \text{PMT} \times [33.999033] \quad \Rightarrow \quad \text{PMT} \approx \$4,797.03$$


**Calculate growing annual deposits**
For deposits that increase by 4% each year, we use the formula for the Future Value of a Growing Annuity and solve for the amount of the first deposit ($C_1$).

$$\text{FV} = C_1 \times \left[ \frac{(1+r)^n - (1+g)^n}{r-g} \right]$$
Where:
* **FV** (Target Goal): $163,094.22

$$\$163,094.22 = C_1 \times \left[ \frac{(1.07)^{18} - (1.04)^{18}}{0.07 - 0.04} \right]$$
$$\$163,094.22 = C_1 \times [45.137192] \quad \Rightarrow \quad C_1 \approx \$3,613.30$$

### Interest Rates

#### Annual Percentage Rate (APR)
* The amount of simple interest earned/paid in one year (the amount of interest earned/paid without the effect of compounding.)
  * Does not reflect the true amount you will earn/pay over one year.
  * Can't use the APR as a discount rate.
  * APR becomes meaningful only if the compounding interval is given.

#### Effective Annual Rate (EAR)
  * The actual amount of interest that will be earned/paid at the end of one year.
$$
(1 + EAR) = (1 + \frac{APR}{k})^k
$$
**Example 2.3**
* If the annual percentage rate (APR) is 6 percent, and compounded quarterly, what is the effective annual rate?
$$
(1 + EAR) = (1 + \frac{0.06}{4})^4
$$
$$
(1 + EAR) = (1 + \frac{0.06}{4})^4 \quad EAR = 6.1364\%
$$

### Example 2.4

Cash Now offers you the following deal: you can write a check for $\$105$ that is dated 14 days in the future, and in return, they give you $\$100$ in cash today. Determine the Annual Percentage Rate (APR) and the Effective Annual Rate (EAR) for this arrangement.

$FV = PV (1 + i)^n \qquad \rightarrow \qquad 105 = 100 (1 + i)^1 \qquad \rightarrow \qquad i=5\%$

$APR = 5\% * (\frac{365}{14}) = 130.3571\%$

$EAR = (1 + \frac{APR}{k})^k - 1 \qquad \rightarrow \qquad EAR = (1 + \frac{130.3571\%}{365/14})^{\frac{365}{14}}-1 \qquad \rightarrow \qquad EAR \approx 256.81\%$

---

```python
import ipywidgets as widgets
from IPython.display import display, clear_output

cash_today = widgets.FloatText(value=100, description='Cash today:')
repayment = widgets.FloatText(value=105, description='Repayment:')
days = widgets.IntText(value=14, description='Days:')
button = widgets.Button(description='Calculate')
output = widgets.Output()

def calculate_apr_ear(_):
    with output:
        clear_output()
        
        pv = cash_today.value
        fv = repayment.value
        n_days = days.value
        
        if pv <= 0 or fv <= 0 or n_days <= 0:
            print("Please enter positive values.")
            return
        
        periodic_rate = (fv - pv) / pv
        periods_per_year = 365 / n_days
        apr = periodic_rate * periods_per_year
        ear = (1 + periodic_rate) ** periods_per_year - 1

        print(f"Periodic rate: {periodic_rate:.4%}")
        print(f"APR: {apr:.4%}")
        print(f"EAR: {ear:.4%}")

button.on_click(calculate_apr_ear)

display(cash_today, repayment, days, button, output)
```

### Interest Rates: Continuous Compounding

* There is no limit to how frequently interest could be paid.
* Compounding every infinitesimal instant is called continuous compounding.

$\qquad FV = PV\left[\left(1 + \frac{r}{k}\right)^k\right]^T$

Let $\, m = \frac{k}{r}$. This implies that $\frac{r}{k} = \frac{1}{m}$ and the exponent $kT$ becomes $(m \cdot r)T$. As the frequency of compounding $k$ approaches infinity, $m$ also approaches infinity.

$$FV = PV \left[\left(1 + \frac{1}{m}\right)^{m \cdot r}\right]^T = PV \left[\left(1 + \frac{1}{m}\right)^m\right]^{rT}$$

The formal definition of the constant $e$ (Euler's number):
$$\lim_{m \to \infty} \left(1 + \frac{1}{m}\right)^m = e$$

Taking the limit of the future value formula as $k$ (and thus $m$) approaches infinity:
$$\lim_{k \to \infty} PV \left(1 + \frac{r}{k}\right)^{kT} = PV \left[ \lim_{m \to \infty} \left(1 + \frac{1}{m}\right)^m \right]^{rT}$$

$$FV = PV \times e^{rT}$$

* What is the present value of an annuity that pays $\$4,500$ per year, with the first cash flow received 6 years from today and the last one received 20 years from today, assuming a discount rate of 8%?

### Exercise 2.6: Solution

The total number of payments in the sequence is:
$$n = (\text{Last payment year} - \text{First payment year}) + 1 \quad \rightarrow \quad n = (20 - 6) + 1 = 15 \text{ payments}$$

#### The Value of the Annuity at Year 5

Since the first payment is at $t=6$, we calculate the value at $t=5$:
$$PV_{t=5} = \text{PMT} \times \left[ \frac{1 - (1+r)^{-n}}{r} \right]$$
$$PV_{t=5} = \$4,500 \times \left[ \frac{1 - (1.08)^{-15}}{0.08} \right] \quad \rightarrow \quad PV_{t=5} = \$4,500 \times 8.55947869 \quad \Rightarrow \quad PV_{t=5} = \$38,517.6541$$

$$PV_{t=0} = \frac{PV_{t=5}}{(1+r)^t}$$
$$PV_{t=0} = \frac{\$38,517.6541}{(1.08)^5} \quad \rightarrow \quad PV_{t=0} = \frac{\$38,517.6541}{1.46932808} \quad \Rightarrow \quad PV_{t=0} \approx \$26,214.47$$

### Exercise 2.7
* What is the value today of a 15-year annuity that pays $\$2,500$ a year? The annuity's first payment occurs three years from today. The annual interest rate is 8 percent for the years 1 through 6, and 12 percent thereafter.

---

### Exercise 2.7: Solution

This problem is unique because the interest rate changes *during* the annuity's payment period. 
    * Payments at $t=3, 4, 5, \text{ and } 6$ occur while the rate is 8%, while payments from $t=7$ through $t=17$ (the remaining 11 payments) occur while the rate is 12%. 
    * To solve this, find the total value of all payments at the point where the interest rate changes ($t=6$) and then discount that total back to today.

#### The Value of all payments at Year 6 ($t=6$)

**Future Value of the first 4 payments ($t=3$ to $t=6$):**
    $$FV_{t=6} = \text{PMT} \times \left[ \frac{(1+r)^n - 1}{r} \right] \quad \Rightarrow \quad FV_{t=6} = \$2,500 \times \left[ \frac{(1.08)^4 - 1}{0.08} \right] \approx \$11,265.28$$

**Present Value of the remaining 11 payments ($t=7$ to $t=17$):**
    $$PV_{t=6} = \text{PMT} \times \left[ \frac{1 - (1+r)^{-n}}{r} \right] \quad \Rightarrow \quad PV_{t=6} = \$2,500 \times \left[ \frac{1 - (1.12)^{-11}}{0.12} \right] \approx \$14,844.25$$

**Total Value at $t=6$:**
    $$\text{Total Value}_{t=6} = \$11,265.28 + \$14,844.25 = \$26,109.53$$

#### Discount the Total Value back to Today ($t=0$)

$$PV_{t=0} = \frac{\text{Total Value}_{t=6}}{(1+r)^n} \quad \Rightarrow \quad PV_{t=0} = \frac{\$26,109.53}{(1.08)^6} \quad \Rightarrow \quad PV_{t=0} = \frac{\$26,109.53}{1.586874} \approx \$16,453.43$$

### Exercise 2.9
You are planning to save for retirement over the next 25 years.
* You will invest $\$1,500$ per month in a diversified equity fund in real dollars and $\$500$ per month in a corporate bond fund in real dollars. 
* The effective annual return of the equity fund is expected to be 9%, and the bond fund will earn 5%. 
* When you retire, you will combine your money into an account with an effective annual return of 6%. The inflation rate over this period is expected to be 2.5%. 
* How much can you withdraw each month from your account in real terms (today's dollars), assuming a withdrawal period of 20 years? What is the nominal dollar amount of your very last withdrawal?

---

### Exercise 2.9: Solution

#### Future Value of Savings in Real Terms
Calculate the real monthly returns for each fund to determine the total future value (FV) of savings at retirement ($t=25$ years), expressed in today's ($t=0$) real dollars.

*   **Real Monthly Equity Rate ($i_e$):**
    $$i_e = \left(\frac{1 + 0.09}{1 + 0.025}\right)^{1/12} - 1 \approx 0.0051369059$$
    $$FV_{equity} = \$1,500 \times \left[ \frac{(1 + 0.0051369059)^{300} - 1}{0.0051369059} \right] \approx \mathbf{\$1,066,169.59}$$

*   **Real Monthly Bond Rate ($i_b$):**
    $$i_b = \left(\frac{1 + 0.05}{1 + 0.025}\right)^{1/12} - 1 \approx 0.0020101469$$
    $$FV_{bond} = \$500 \times \left[ \frac{(1 + 0.0020101469)^{300} - 1}{0.0020101469} \right] \approx \mathbf{\$205,598.90}$$

**Total Real Savings at Retirement ($S$):** 
$\$1,066,169.59 + \$205,598.90 = \mathbf{\$1,271,768.49}$

#### Real Monthly Retirement Withdrawal
Calculate the constant real monthly withdrawal ($W$) for 20 years (240 months).

*   **Real Monthly Retirement Rate ($i_r$):**
    $$i_r = \left(\frac{1 + 0.06}{1 + 0.025}\right)^{1/12} - 1 \approx 0.0028019428$$

$$W = S \times \left[ \frac{i_r}{1 - (1 + i_r)^{-240}} \right] \quad \Rightarrow \quad W = \$1,271,768.49 \times \left[ \frac{0.0028019428}{1 - (1.0028019428)^{-240}} \right] \approx \mathbf{\$7,286.10}$$

#### Nominal Value of Last Withdrawal
The last withdrawal occurs at $t=45$ years (25 years saving + 20 years retirement). We inflate the real amount to find its nominal value.

$$W_{last} = W \times (1 + \text{inflation})^{45} \quad \Rightarrow \quad W_{last} = \$7,286.10 \times (1.025)^{45} \approx \mathbf{\$22,134.45}$$


---

```python
import math

# Real annual equity
real_annual_eq = (1 + 0.09) / (1 + 0.025) - 1
print(f"Real annual equity: {real_annual_eq}")

# Monthly equity
i_e = (1 + real_annual_eq)**(1/12) - 1
print(f"i_e: {i_e}")

# FV equity
fv_factor_eq = ((1 + i_e)**300 - 1) / i_e
fv_equity = 1500 * fv_factor_eq
print(f"FV equity: {fv_equity}")

# Real annual bond
real_annual_bond = (1 + 0.05) / (1 + 0.025) - 1
print(f"Real annual bond: {real_annual_bond}")

# Monthly bond
i_b = (1 + real_annual_bond)**(1/12) - 1
print(f"i_b: {i_b}")

# FV bond
fv_factor_bond = ((1 + i_b)**300 - 1) / i_b
fv_bond = 500 * fv_factor_bond
print(f"FV bond: {fv_bond}")

# Total S
total_s = fv_equity + fv_bond
print(f"Total S: {total_s}")

# Real monthly retirement
real_annual_ret = (1 + 0.06) / (1 + 0.025) - 1
i_r = (1 + real_annual_ret)**(1/12) - 1
print(f"i_r: {i_r}")

# Withdrawal factor
denom = 1 - (1 + i_r)**(-240)
w_factor = i_r / denom
W = total_s * w_factor
print(f"W: {W}")

# Last nominal
last_w = W * (1.025)**45
print(f"Last nominal: {last_w}")
```
