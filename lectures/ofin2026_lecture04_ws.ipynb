### Capital Budgeting

**Capital Budget**
*   Lists the investments that a company plans to undertake.

**Capital Budgeting**
*   Process used to analyze alternative investments.
*   Ultimate goal is to determine the effect of the decision on the firm's cash flows.

**Incremental Earnings**
*   The amount by which the firm's earnings are expected to change as a result of the investment decision.

**Capital Budgeting Process**
*   Revenue and cost estimates
*   Incremental earnings forecast
*   Free cash flow and NPV
*   Choose among alternatives

### Capital Budgeting: Discount cash flows, not profits

*   Accounting earnings are often drastically different from actual cash flows. In capital budgeting, you must always discount cash flows, not accounting net income.

*   Consider a new server installation that costs $12,000 upfront (Year 0). It is expected to generate cash revenues of $8,000 in Year 1 and $6,000 in Year 2. The accounting department depreciates the hardware straight-line over the two years.

    |                     | Year 1   | Year 2   |
    | :------------------ | :------- | :------- |
    | Cash Revenue        | +$8,000  | +$6,000  |
    | Less Depreciation   | -$6,000  | -$6,000  |
    | Accounting Income   | +$2,000  | $0       |

    *   If we mistakenly discount the accounting income using a 10% cost of capital:
    $$ NPV = \frac{\$2,000}{1.10^1} + \frac{\$0}{1.10^2} = \$1,818.18 \ \ \ <-- \text{WRONG!} $$

    *   Instead, we must discount the actual cash flows generated by the project:
    $$ NPV = -\$12,000 + \frac{\$8,000}{1.10^1} + \frac{\$6,000}{1.10^2} $$
    $$ NPV = -\$12,000 + \$7,272.73 + \$4,958.68 = \$231.40 \ \ \ <-- \text{CORRECT!} $$

### Capital Budgeting: Working capital matters

*   Investment in net working capital, such as accounts receivable, significantly shifts the timing of cash flows. The moment revenue is recognized in accounting is rarely the moment cash actually enters the bank account.

*   A marketing agency takes on a major campaign. They must pay their media buyers $20,000 at t=1. The client is billed $35,000 when the campaign launches at t=1, but the invoice allows for net-365 payment terms, meaning the cash does not arrive until t=2.

    **Accounting perspective (accrual basis):**
    |                     | Amount($) | Timing |
    | :------------------ | ------: | -----: |
    | Revenue Recognized  | 35,000 | t=1    |
    | Less Cost of Goods  | - 20,000 | t=1    |
    | Accounting Income   | 15,000 | t=1    |

    **Cash Flow perspective:**
    |                            | t=0 | t=1 | t=2 |
    | :------------------------- | -------: | -------: | -------: |
    | **Components of CF:**      |          |          |          |
    | Cash Outlay (Media Buyers) |      0   | -20,000  |       0  |
    | Cash Inflow (Client Pmt)   |      0   |      0   | +35,000  |
    | Net Cash flow              |      0   | -20,000  | +35,000  |
       

* The timing difference between cash flow and accounting income can significantly impact project valuation. 
* **Always track the actual cash inflows and outflows.**

### Capital Budgeting: Discount incremental cash flows

*   The value of a project depends on all the additional cash flows that follow from project acceptance.
*   Include all incidental effects
    *   Introduction of a new product will almost certainly cut into sales of existing products (Product Cannibalization).
*   Do not confuse average with incremental payoffs
    *   The incremental NPV from investing in a loser may be strongly positive.
    *   Also, it does not always make sense to throw good money after good.
*   Forecast product sales but also recognize after-sales cash flows.
*   Include opportunity costs.
*   Forget sunk costs.
    *   They are past and irreversible outflows.
*   Beware of allocated overhead costs.
    *   In investment appraisal we should include only the extra expenses that would result from the project.
*   Remember salvage value
    *   When the project comes to an end, you may be able to sell the plant and equipment or redeploy the assets elsewhere in the business.

### Capital Budgeting: Discount Incremental Cash Flows

Delta Appliances, a manufacturer of high-end refrigerators, is considering introducing a new smart-connectivity model. The firm anticipates that annual sales of the new refrigerator will be $80 million. While many sales will attract new customers, Delta estimates that 25% of these sales will come from existing customers who opt for the smart model instead of the company's current standard model.

*   Assume customers spend roughly the same amount on either model. What level of incremental sales is associated with introducing the new smart-connectivity refrigerator?

*   Now, suppose Delta's market research indicates that 40% of the customers who would potentially switch from the original refrigerator to the new one would likely switch to a competitor's smart model if Delta doesn't launch its own. What is the level of incremental sales in this case?

---

### Capital Budgeting: Discount Incremental Cash Flows

#### Basic Incremental Sales

$$\text{Incremental Sales} = \text{Total New Product Sales} \times (1 - \text{Cannibalization Rate})$$

$$\text{Incremental Sales} = \$80 mio \times (1 – 25\%) = \$80 mio \times 0.75 = \$60 mi0$$

#### Lost Sales Avoided
* Some cannibalized sales are not actually "lost" because those customers would have left Delta entirely for a competitor if the new product were not available. 

$$\text{Incremental Sales} = \text{Total New Product Sales} \times [1 – \text{Cannibalization Rate} \times \text{\% Customers Staying with Original Model})]$$

1. Percentage of customers who would have stayed with the original model = 1 - 40% = 60%

2. Effective cannibalization rate = 25% $\times$ 60% = 15%

Incremental Sales = \$80 mio $\times$ (1 – 15%) = \$68 mio

Using the lost sales avoided method:

Incremental Sales = Total Sales $\times$ (1 - Cannibalization Rate + Lost Sales Avoided)

Incremental Sales = \$80 mio $\times$ (1 - 0.25 + (0.25 $\times$ 0.40)) = \$68 million

### Capital Budgeting: Discount Incremental Cash Flows

Theta Brewing Systems currently sells its "Barista-X" commercial espresso machine for $\$1,500$. To compete with new entry-level competitors, the company is considering lowering the price to $\$1,200$ for all units sold next year. The cost of goods sold (COGS) for each machine is $\$850$. Current annual sales volume is 2,000 units.

*   Suppose that if Theta Brewing Systems drops the price to $\$1,200$ immediately, it expects annual sales volume to increase by 50% to 3,000 units. What is the incremental impact on this year’s EBIT resulting from this price reduction?

$$\text{EBIT} = \text{Sales Volume} \times (\text{Price} - \text{COGS})$$

*   **Original EBIT:** $2,000 \text{ units} \times (\$1,500 - \$850) = 2,000 \times \$650 = \$1,300,000$

*   **New EBIT:** $3,000 \text{ units} \times (\$1,200 - \$850) = 3,000 \times \$350 = \$1,050,000$

*   **Change in EBIT:** $\$1,050,000 - \$1,300,000 = -\$250,000$

### Capital Budgeting: Treat inflation consistently

Discount nominal cash flows using the nominal discount rate, and real cash flows using the real discount rate.

Suppose Omega Manufacturing typically forecasts project cash flows in nominal terms and discounts them at a 10% nominal cost of capital. For a new industrial conveyor system, initial estimates are provided in today's dollars (real terms), and inflation is projected to be 4% per year.

| Real Cash Flows ($ thousands) | $C_0$ | $C_1$ | $C_2$ | $C_3$ |
| :---------------------------- | :---: | :---: | :---: | :---: |
| Value                         | -200  | +90   | +100  | +60   |

Real discount rate $(r_r) = \frac{1 + \text{Nominal discount rate}}{1 + \text{Inflation rate}} - 1 \quad \rightarrow \quad r_r = \frac{1 + 0.10}{1 + 0.04} - 1 \approx 5.7692\% $

Two Approaches to Calculate NPV:

1. Discount Real Cash Flows at the Real Discount Rate (5.7692%)

2. Convert Real Cash Flows to Nominal, then Discount at the Nominal Discount Rate (10.00%)

| t | Real CF (\$k) | Nominal CF (\$k) [Real $\times$ (1+i)^t] | PV Real CF @ 5.7692% ($k) | PV Nominal CF @ 10.00% ($k) |
| :-: | -----------: | -----------------------------: | ----------------------: | ------------------------: |
| 0 | -200.00 | -200.00 | -200.00 | -200.00 |
| 1 | 90.00 | 93.60 | 85.09 | 85.09 |
| 2 | 100.00 | 108.16 | 89.39 | 89.39 |
| 3 | 60.00 | 67.49 | 50.71 | 50.71 |
| | | | NPV = 25.19 | NPV = 25.19 |

### Capital Budgeting

#### Separate investment and financing decisions
*   **Separation Principle:** When securities are fairly priced, the NPV of financing is zero, and so the NPV of an investment is independent of how it is financed.

#### Remember to Deduct Taxes
*   Taxes are an expense just like wages and raw materials.

*   Therefore, cash flows should be estimated on an after-tax basis.

*   Subtract cash outflows for taxes from pretax cash flows and discount the net amount.

### Determining Free Cash Flow and NPV
*   Earnings are just an accounting measure of the firm's performance.
    *   Earnings include non-cash charges, such as depreciation, but do not include the CAPEX.
*   The incremental effect of a project on the firm's available cash, separate from any financing decisions, is the project's free cash flow.
*   FCF calculation starts with Unlevered Net Income (UNI) and adjusts for non-cash expenses and cash flows related to investments.
    *   Add back non-cash expenses (like depreciation).
    *   Subtract capital expenditures (CAPEX).
    *   Subtract "increases in net working capital (NWC) or add decreases in NWC.

---

### Determining Free Cash Flow and NPV
*   Most projects require investments in Net Working Capital (NWC) to support operations. 
    * NWC represents the difference between short-term assets and short-term liabilities needed for the project.
    
    $ \text{Net Working Capital} = \text{Current Assets} - \text{Current Liabilities} \quad \rightarrow \quad NWC = (\text{Cash}_\text{req} + \text{Inventory} + \text{Receivables}) - (\text{Payables})$

*   **Assumptions:**
    *   Titan-X requires no incremental cash balances beyond normal corporate levels.
    *   Inventory levels are managed efficiently, requiring no significant incremental investment relative to sales.
    *   Accounts Receivable are projected to be **10%** of annual sales.
    *   Accounts Payable are projected to be **20%** of the annual Cost of Goods Sold (COGS).

### Calculating Free Cash Flow Directly

1.  **Using Unlevered Net Income:**
    $$ \text{Free Cash Flow} = \overbrace{(\text{Revenues} – \text{Costs} - \text{Depreciation}) \times (1 – \tau_c)}^{\text{Unlevered Net Income}} + \text{Depreciation} – \text{CapEx} – \Delta \text{NWC} $$
    *   Where "Costs" represent operating costs excluding depreciation (e.g., COGS + SG&A).
    *   $\tau_c$ is the corporate tax rate.

2.  **Using operating cash flow before depreciation tax shield:**
    $$ \text{Free Cash Flow} = (\text{Revenues} – \text{Costs}) \times (1 – \tau_c) – \text{CapEx} – \Delta \text{NWC} + (\tau_c \times \text{Depreciation}) $$
    *   This formula calculates the after-tax cash flow from operations as if there were no depreciation, then separately adds the tax savings provided by the depreciation deduction (the Depreciation Tax Shield, $\tau_c \times \text{Depreciation}$).

3.  **Starting from EBIT:**
    $$ \text{Free Cash Flow} = \overbrace{\text{EBIT} \times (1 – \tau_c)}^{\text{Unlevered Net Income}} + \text{Depreciation} – \text{CapEx} – \Delta \text{NWC} $$

### Present Value Calculation

$$ PV(FCF_t) = \frac{FCF_t}{(1+r)^t} = FCF_t \times \overbrace{\frac{1}{(1+r)^t}}^{\text{Discount Factor for Year t}} $$

### Calculating Operating Cash Flow (OCF)

*   A specialized software firm, Omega Solutions, estimates that its new cybersecurity consulting service will generate annual sales of $800,000. The associated operating costs (salaries, licensing, cloud infrastructure) are expected to be $480,000 per year, and the annual depreciation expense on the new server hardware and security appliances is $80,000. If the firm's tax rate is 21%, what is the expected annual Operating Cash Flow (OCF) from this new service?

    **Method 1: Tax Shield Approach**
    
    $ OCF = (\text{Revenues} – \text{Costs}) \times (1 - \tau_c) + (\tau_c \times \text{Depreciation}) $
    
    $ OCF = (\$800,000 – \$480,000) \times (1 – 0.21) + (0.21 \times \$80,000) $
    
    $ OCF = (\$320,000) \times (0.79) + \$16,800 \rightarrow OCF = \$269,600 $

    **Method 2: Unlevered Net Income + Depreciation**
    
    $ OCF = \overbrace{(\text{Revenues} – \text{Costs} - \text{Depreciation}) \times (1 − \tau_c)}^{\text{Unlevered Net Income}} + \text{Depreciation} $
    
    $ \text{Unlevered NI} = (\$800,000 – \$480,000 – \$80,000) \times (1 – 0.21) $
    
    $ \text{Unlevered NI} = (\$240,000) \times (0.79) = \$189,600 $

    $ OCF = \text{Unlevered Net Income} + \text{Depreciation} $
    
    $ OCF = \$189,600 + \$80,000 = \$269,600 $

### Break-even Analysis
*   Sigma Heavy Industries is evaluating the development of a new heavy-duty industrial pump, called Titan-X.
    *   The pump aims to provide high-efficiency fluid transfer for mining operations using a proprietary magnetic drive. Sigma has already completed a $\$600,000$ market analysis study which suggested strong demand.

    *   Initial core technology development and engineering costs are estimated at $\$5$ million.
    *   Manufacturing will be handled in-house, with an estimated variable production cost of $\$1,000$ per unit.

    *   An investment of $\$15$ million in new manufacturing equipment is required today ($t=0$) and will be depreciated straight-line over its $5$-year useful life.

    *   The product development and setup will take one year (Year 0), with sales commencing at the beginning of Year 1. 
    *   Titan-X requires no incremental cash balances. Accounts receivable are projected to be $10\%$ of annual sales. Accounts payable are projected to be $20\%$ of the annual Cost of Goods Sold (COGS).
    *   The cost of capital for the project is $10\%$, and the corporate tax rate is $30\%$.

### Break-even analysis – Titan-X

- Initial investment (Year 0): $\quad FCF_0 = -(R\&D \times (1 - Tax) + CapEx) $
- Year 1 FCF: $\qquad \qquad \qquad \quad FCF_1(Q, P, VC) = OCF_1(Q, P, VC) - NWC_1(Q, P, VC)$
- Years 2–4 FCF: $\qquad \qquad \qquad FCF_{2-4}(Q, P, VC) = OCF(Q, P, VC)$
- Year 5 FCF: $\qquad \qquad \qquad \quad FCF_5(Q, P, VC) = OCF_5(Q, P, VC) + NWC_1(Q, P, VC)$

- Operating Cash Flow (OCF): $\quad OCF(Q, P, VC) = \left[ (Q \cdot P - Q \cdot VC - SG\&A - Depreciation) \cdot (1 - Tax) \right] + Depreciation$

- Net Working Capital (NWC): $\quad NWC_1(Q, P, VC) = 0.10 \cdot (P \cdot Q) - 0.20 \cdot (VC \cdot Q)$

**Objective Function for Break-Even Calculation:**

The break-even point occurs when the project's Net Present Value (NPV) is zero:

$NPV(Q,P,VC,r) = FCF_0 + \frac{FCF_1(Q,P,VC)}{(1+r)^1} + \frac{FCF_2(Q,P,VC)}{(1+r)^2} + \frac{FCF_3(Q,P,VC)}{(1+r)^3} + \frac{FCF_4(Q,P,VC)}{(1+r)^4} + \frac{FCF_5(Q,P,VC)}{(1+r)^5}=0$

**Parameter**           | **Break-Even Level**
-------------------------|-------------------------------
Units sold ($Q$)         | 7,010 units per year  
Wholesale price ($P$)    | \$2,048.27 per unit  
Cost of goods ($VC$)     | \$1,456.74 per unit  
Cost of capital ($r$)    | 30.24% (IRR)

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve

# =============================================================================
# CONSTANTS & BASE CASE PARAMETERS (TITAN-X PROJECT)
# =============================================================================
BASE_UNITS = 10000
BASE_PRICE = 2500.0
BASE_COGS = 1000.0
FIXED_SGA = 4000
ANNUAL_DEPR = 3000
TAX_RATE = 0.30
CAPEX_INITIAL = 15000
RD_INITIAL = 8000
BASE_DISCOUNT_RATE = 0.10

# =============================================================================
# CORE MODEL FUNCTIONS
# =============================================================================
def calculate_fcfs(units=BASE_UNITS, price=BASE_PRICE, cogs=BASE_COGS):
    sales = units * price / 1000
    variable_costs = units * cogs / 1000

    ebit = sales - variable_costs - FIXED_SGA - ANNUAL_DEPR
    tax = ebit * TAX_RATE
    net_income = ebit - tax
    ocf = net_income + ANNUAL_DEPR

    nwc_requirement = 0.10 * sales - 0.20 * variable_costs

    fcfs = np.zeros(6)
    fcfs[0] = -(RD_INITIAL * (1 - TAX_RATE) + CAPEX_INITIAL)
    fcfs[1] = ocf - nwc_requirement
    fcfs[2:5] = ocf
    fcfs[5] = ocf + nwc_requirement
    return fcfs

def calculate_npv(fcfs, rate=BASE_DISCOUNT_RATE):
    return sum(fcfs[t] / (1 + rate)**t for t in range(len(fcfs)))

# =============================================================================
# NUMERICAL ROOT SOLVER (FSOLVE)
# =============================================================================
def obj_q(q): return calculate_npv(calculate_fcfs(units=float(np.asarray(q).flat[0])))
def obj_p(p): return calculate_npv(calculate_fcfs(price=float(np.asarray(p).flat[0])))
def obj_c(c): return calculate_npv(calculate_fcfs(cogs=float(np.asarray(c).flat[0])))
def obj_r(r): return calculate_npv(calculate_fcfs(), rate=float(np.asarray(r).flat[0]))

be_units = fsolve(obj_q, BASE_UNITS)[0]
be_price = fsolve(obj_p, BASE_PRICE)[0]
be_cogs  = fsolve(obj_c, BASE_COGS)[0]
be_irr   = fsolve(obj_r, BASE_DISCOUNT_RATE)[0]

# =============================================================================
# VISUALIZATION (2x2 PANEL)
# =============================================================================
fig, axs = plt.subplots(2, 2, figsize=(14, 10))
plt.subplots_adjust(hspace=0.35, wspace=0.25)

# --- Plot 1: NPV vs. Units Sold ---
q_range = np.linspace(be_units * 0.5, be_units * 1.5, 100)
axs[0, 0].plot(q_range, [obj_q(q) for q in q_range], lw=2, color='tab:blue')
axs[0, 0].set_title('NPV Sensitivity: Units Sold', fontsize=12)
axs[0, 0].set_xlabel('Annual Units Sold')
axs[0, 0].scatter(be_units, 0, color='red', zorder=5)
axs[0, 0].annotate(f'Break-even: {be_units:.0f}', (be_units, 1000), color='red', weight='bold')

# --- Plot 2: NPV vs. Wholesale Price ---
p_range = np.linspace(be_price * 0.7, be_price * 1.3, 100)
axs[0, 1].plot(p_range, [obj_p(p) for p in p_range], lw=2, color='tab:green')
axs[0, 1].set_title('NPV Sensitivity: Price', fontsize=12)
axs[0, 1].set_xlabel('Price per Unit ($)')
axs[0, 1].scatter(be_price, 0, color='red', zorder=5)
axs[0, 1].annotate(f'Break-even: ${be_price:.2f}', (be_price, 1000), color='red', weight='bold')

# --- Plot 3: NPV vs. Unit Variable Cost (COGS) ---
c_range = np.linspace(be_cogs * 0.5, be_cogs * 1.5, 100)
axs[1, 0].plot(c_range, [obj_c(c) for c in c_range], lw=2, color='tab:orange')
axs[1, 0].set_title('NPV Sensitivity: Variable Cost', fontsize=12)
axs[1, 0].set_xlabel('COGS per Unit ($)')
axs[1, 0].scatter(be_cogs, 0, color='red', zorder=5)
axs[1, 0].annotate(f'Break-even: ${be_cogs:.2f}', (be_cogs, 1000), color='red', weight='bold')

# --- Plot 4: NPV vs. Cost of Capital (IRR) ---
r_range = np.linspace(0.05, be_irr * 1.5, 100)
axs[1, 1].plot(r_range * 100, [obj_r(r) for r in r_range], lw=2, color='tab:red')
axs[1, 1].set_title('NPV Sensitivity: Discount Rate (IRR)', fontsize=12)
axs[1, 1].set_xlabel('Discount Rate (%)')
axs[1, 1].scatter(be_irr * 100, 0, color='black', zorder=5)
axs[1, 1].annotate(f'IRR: {be_irr*100:.2f}%', (be_irr*100, 1000), weight='bold')

# Global Plot Formatting
for ax in axs.flat:
    ax.axhline(0, color='black', lw=1, ls='--')
    ax.grid(alpha=0.3)
    ax.set_ylabel('NPV ($000)')

plt.suptitle('Titan-X Project: Break-Even Analysis', fontsize=16)
plt.show()

# =============================================================================
# CONSOLE SUMMARY
# =============================================================================
print(f"--- Titan-X Break-Even Analysis Summary ---")
print(f"Base Case NPV:  ${calculate_npv(calculate_fcfs()):,.2f}")
print(f"Project IRR:    {be_irr*100:.4f}%")
print(f"Break-even Q:   {be_units:,.2f} units")
print(f"Break-even P:   ${be_price:,.2f}")
print(f"Break-even VC:  ${be_cogs:,.2f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# =============================================================================
# CONSTANTS & SCENARIO PARAMETERS
# =============================================================================
units_initial = 50000
price_initial = 450
cogs_initial = 180
discount_initial = 0.12

FIXED_SGA = 3500
ANNUAL_DEPR = 2000
TAX_RATE = 0.35
CAPEX_INITIAL = 10000
RD_INITIAL = 15500

# NWC rule: 10% of Sales - 20% of COGS (dynamic, same convention as Titan-X)
def calculate_nwc(units, price, cogs):
    sales = units * price / 1000
    variable_costs = units * cogs / 1000
    return 0.10 * sales - 0.20 * variable_costs

scenarios = {
    'Units Sold':      [30000, units_initial, 75000],
    'Sale Price':      [250,   price_initial,  575],
    'Cost of Goods':   [300,   cogs_initial,   120],
    'Cost of Capital': [0.18,  discount_initial, 0.10]
}

# =============================================================================
# CORE MODEL FUNCTIONS
# =============================================================================
def calculate_fcfs(units=units_initial, price=price_initial, cogs=cogs_initial):
    sales = units * price / 1000
    cost_of_goods = units * cogs / 1000
    ebit = sales - cost_of_goods - FIXED_SGA - ANNUAL_DEPR
    net_income = ebit - ebit * TAX_RATE
    ocf = net_income + ANNUAL_DEPR

    nwc = calculate_nwc(units, price, cogs)

    fcfs = np.zeros(6)
    fcfs[0] = -(RD_INITIAL * (1 - TAX_RATE) + CAPEX_INITIAL)
    fcfs[1] = ocf - nwc        # NWC outflow in year 1
    fcfs[2:5] = ocf
    fcfs[5] = ocf + nwc        # NWC recovery in year 5
    return fcfs

def calculate_npv(fcfs, rate=discount_initial):
    return sum(fcfs[t] / (1 + rate)**t for t in range(len(fcfs)))

# =============================================================================
# SCENARIO NPV COMPUTATION
# =============================================================================
base_npv = calculate_npv(calculate_fcfs())

npv_ranges = {}
for param, (worst, base, best) in scenarios.items():
    if param == 'Units Sold':
        low  = calculate_npv(calculate_fcfs(units=worst))
        high = calculate_npv(calculate_fcfs(units=best))
    elif param == 'Sale Price':
        low  = calculate_npv(calculate_fcfs(price=worst))
        high = calculate_npv(calculate_fcfs(price=best))
    elif param == 'Cost of Goods':
        low  = calculate_npv(calculate_fcfs(cogs=worst))
        high = calculate_npv(calculate_fcfs(cogs=best))
    elif param == 'Cost of Capital':
        low  = calculate_npv(calculate_fcfs(), rate=worst)
        high = calculate_npv(calculate_fcfs(), rate=best)
    npv_ranges[param] = [low, high]

# =============================================================================
# SORT BY IMPACT WIDTH (proper tornado ordering)
# =============================================================================
sorted_ranges = dict(sorted(npv_ranges.items(), key=lambda x: abs(x[1][1] - x[1][0]), reverse=False))

# =============================================================================
# VISUALIZATION — TORNADO DIAGRAM
# =============================================================================
labels = list(sorted_ranges.keys())
lows  = [sorted_ranges[l][0] / 1000 for l in labels]
highs = [sorted_ranges[l][1] / 1000 for l in labels]
base_m = base_npv / 1000

fig, ax = plt.subplots(figsize=(12, 6))
for i, label in enumerate(labels):
    ax.barh(i, highs[i] - lows[i], left=lows[i],
            color='green' if highs[i] > lows[i] else 'red', alpha=0.6, height=0.5)

    worst_val, _, best_val = scenarios[label]
    if label == 'Cost of Capital':
        w_str = f"{worst_val*100:.0f}%"
        b_str = f"{best_val*100:.0f}%"
    elif label in ('Sale Price', 'Cost of Goods'):
        w_str = f"${worst_val:,.0f}"
        b_str = f"${best_val:,.0f}"
    else:
        w_str = f"{worst_val:,.0f} units"
        b_str = f"{best_val:,.0f} units"

    ax.text(lows[i] + 0.2, i, f"Worst\n{w_str}\nNPV: ${lows[i]*1000:,.0f}",
            ha='left', va='center', fontsize=7.5, color='darkred')
    ax.text(highs[i] - 0.2, i, f"Best\n{b_str}\nNPV: ${highs[i]*1000:,.0f}",
            ha='right', va='center', fontsize=7.5, color='darkgreen')

ax.set_yticks(range(len(labels)))
ax.set_yticklabels(labels)
ax.axvline(base_m, color='black', linestyle='--', label=f'Base NPV: ${base_npv:,.0f}')
ax.set_xlabel('Project NPV ($000)')
ax.set_title('Sensitivity Analysis — Tornado Diagram')
ax.legend(loc='lower right', fontsize=9)
ax.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# =============================================================================
# CONSOLE SUMMARY
# =============================================================================
print(f"--- Tornado Analysis Summary ---")
print(f"Base Case NPV: ${base_npv:,.2f}\n")
print(f"{'Parameter':<20} {'Worst Input':<18} {'Worst NPV':>14}   {'Best Input':<18} {'Best NPV':>14}   {'Range':>14}")
print("-" * 100)
for param, (low, high) in sorted_ranges.items():
    worst_val, _, best_val = scenarios[param]
    if param == 'Cost of Capital':
        w_str = f"{worst_val*100:.0f}%"
        b_str = f"{best_val*100:.0f}%"
    elif param in ('Sale Price', 'Cost of Goods'):
        w_str = f"${worst_val:,.0f}"
        b_str = f"${best_val:,.0f}"
    else:
        w_str = f"{worst_val:,.0f} units"
        b_str = f"{best_val:,.0f} units"
    print(f"{param:<20} {w_str:<18} ${low:>12,.2f}   {b_str:<18} ${high:>12,.2f}   ${high-low:>12,.2f}")
```
