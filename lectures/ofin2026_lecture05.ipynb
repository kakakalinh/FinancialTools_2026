### Bond Valuation
* Fixed-income securities markets (mostly) consist of long-term debt instruments.
* Because most of these instruments provide a regular or formula-based cash flow, they are called fixed-income securities.
* The main examples of fixed-income securities:
    * Treasury bills and bonds,
    * Corporate bills and bonds,
    * Local government bills and bonds,
    * Mortgage-backed securities,
    * Asset-backed securities

* Fixed income securities may be classified according to some of their characteristics:
    * Recallable bonds (Call option of the issuer)
    * Redeemable bonds (Put option of the bondholders)
    * Convertible bonds (i.e. convertible to stocks)
    * Preferred stocks (Both stock and fixed income features)
    * Indexed bonds (Oil prices, inflation, GNP, etc.)
    * Impact bonds (Socially responsible investments)
    * Green and blue bonds (Focused on sustainability and environmental protection)

---

### Bond Valuation
* **A bill vs. a bond.**
    * Bills are short-term instruments whereas bonds are issued for longer terms.
    * There are no coupon payments on bills, only a single payment (the face value of the bond) at maturity.
    * In bonds, there are coupon payments made in pre-determined intervals, and the nominal amount of the bond is paid at maturity.
        * Coupon payments may be fixed or variable.
        * Fixed coupon payments are set at the bond issuance.
        * Variable interest, on the other hand, is determined in connection to a specific reference. This reference interest rate may be based on current market rates, or it may be based on inflation or a similar indicator.


* **What is the main difference between government, corporate and local government bonds?**
    * Divergence of risk profiles.
    * Regarding the risk profiles, liquidity and default risks are the main determinants.
        * For an asset to be considered a liquid asset, it must be able to be converted into cash when needed without significant loss of value.
            * Liquidity is closely related to the market depth of the security. It may differ across an issuer's various issuances, and it may also differ across issuers.
        * Default risk refers to the risk of the issuer not making the payments on the debt on time. It is present in all debt instruments at varying levels and directly affects the return that investors demand.

### Bond Valuation
* Pricing Fixed Income Securities
    * The price of any financial instrument is equal to the present value of the cash flows expected from the instrument.
    * Therefore, setting the price requires the estimation of expected cash flows and estimation of expected return.
    * The first step in determining the price of a fixed-income security is to estimate its cash flows.
    * Cash flows for a bond that does not have an embedded option consist of periodic coupon interest payments until maturity and face/par/nominal value at maturity.
    * Bills or zero-coupon bonds are a special class of bonds with no coupon payments, only face value.
* A bond with fixed coupon payments may be considered an annuity for which payments are made at the end of each period.
* Unlike annuities, bonds have an additional payment, face value, made at maturity.
$$ PV = CF \left[ \frac{1 - \frac{1}{(1+i)^T}}{i} \right] + \frac{Face (Par)Value}{(1+i)^T} $$
* The cash flows for a coupon-paying bond may be written in the following form:
$$ P = \frac{C}{(1+r)^1} + \frac{C}{(1+r)^2} + \frac{C}{(1+r)^3} + \dots + \frac{C}{(1+r)^n} + \frac{FV}{(1+r)^n} $$
$$ P = \sum_{t=1}^{n} \frac{C}{(1+r)^t} + \frac{FV}{(1 + r)^n} $$
where P is the price of the bond at t=0. C represents fixed coupon payments to be made at the end of each period from t=1 to t=n. FV is the face or par value of the bond to be repaid at maturity. Market interest rate is denominated by r, which may also be interpreted as the required rate of return.

### Bond Valuation
**What is the price of a 5-year bond at the issuance date that has a face value of €10,000 and pays annual coupons of 12% when the market interest rate is at 10%?**

$ P = \frac{1,200}{(1 + 0.10)^1} + \frac{1,200}{(1 + 0.10)^2} + \frac{1,200}{(1 + 0.10)^3} + \frac{1,200}{(1 + 0.10)^4} + \frac{1,200}{(1 + 0.10)^5} + \frac{10,000}{(1 + 0.10)^5} $

$ P = 1,090.91+ 991.74 + 901.58 + 819.62 + 745.11 + 6,209.21 = 10,758.17 $

**What is the price of a 5-year bond that has a nominal value of €10,000 and pays annual coupons of 12% on the issuance date when the market interest rate is at 15%?**

$ P = \frac{1,200}{(1+0.15)^1} + \frac{1,200}{(1 + 0.15)^2} + \frac{1,200}{(1 + 0.15)^3} + \frac{1,200}{(1 + 0.15)^4} + \frac{1,200}{(1 + 0.15)^5} + \frac{10,000}{(1 + 0.15)^5} $

$ P = 1,043.48 + 907.37 + 789.02 + 686.10 + 596.61 + 4,973.77 = 8,996.35 $

$ P = C \left[ \frac{1 - \frac{1}{(1 + i)^T}}{i} \right] + \frac{Face (Par)Value}{(1 + i)^T} $

$ P = 1,200 \left[ \frac{1 - \frac{1}{(1 + 0.15)^5}}{0.15} \right] + \frac{10,000}{(1 + 0.15)^5} = 4,022.59 + 4,973.77 = 8,996.36 $

### Bond Valuation
* Up to this point, we assumed that the coupon payments are made annually. However, it may also be the case that the coupons are paid more frequently, i.e. semi-annually or quarterly.
* What is the price of a 5-year bond that has a nominal value of €10,000, and an annual coupon rate of 12% which is paid semi-annually? Assume that the market interest rate is at 10%.
$$ P = \left( \sum_{t=1}^{10} \frac{600}{(1 + (0.10/2))^{5*2}} \right) + \frac{10,000}{(1 + 0.10/2)^{5*2}} = 10,772.17 $$
* What is the price if coupon payments are paid quarterly?
$$ P = \left( \sum_{t=1}^{20} \frac{300}{(1 + (0.10/4))^{5*4}} \right) + \frac{10,000}{(1 + 0.10/4)^{5*4}} = 10,779.46 $$


**Coupon Frequency and Bond Price**

| Coupon Frequency | Coupon Rate | Market Interest Rate | Face Value | Bond Price |
|:------------------:|:-------------:|----------------------:|------------:|------------:|
| 1 year           | 12%         | 10%                  | 10,000.00  | 10,758.15  |
| 6 months         | 6%          | 10%                  | 10,000.00  | 10,772.17  |
| 3 months         | 3%          | 10%                  | 10,000.00  | 10,779.46  |

**The higher the coupon payment frequency, the higher the bond price when all other variables are constant.**

```python
from IPython.display import display, clear_output
import ipywidgets as widgets
import numpy as np

def calculate_bond_price(face_value, coupon_rate_percent, market_rate_percent, maturity, frequency):
    coupon_rate = coupon_rate_percent / 100
    market_rate = market_rate_percent / 100
    periods = int(maturity * frequency)
    coupon_payment = face_value * coupon_rate / frequency
    discount_factors = [(1 + market_rate / frequency) ** t for t in range(1, periods + 1)]
    present_value_coupons = sum(coupon_payment / df for df in discount_factors)
    present_value_face = face_value / discount_factors[-1]
    return round(present_value_coupons + present_value_face, 2)

# Layout settings for label width and input box size
label_style = {'description_width': '160px'}
input_layout = widgets.Layout(width='300px')

# Input widgets
face_value_input = widgets.FloatText(value=10000, description='Face Value:', style=label_style, layout=input_layout)
coupon_rate_input = widgets.FloatText(value=12.0, description='Coupon Rate (%):', style=label_style, layout=input_layout)
market_rate_input = widgets.FloatText(value=10.0, description='Market Rate (%):', style=label_style, layout=input_layout)
maturity_input = widgets.IntSlider(value=5, min=1, max=30, description='Maturity (years):', style=label_style, layout=input_layout)
frequency_input = widgets.Dropdown(options=[1, 2, 4], value=1, description='Payments/year:', style=label_style, layout=input_layout)

output = widgets.Output()

def on_calculate_button_clicked(b):
    with output:
        clear_output()
        price = calculate_bond_price(
            face_value_input.value,
            coupon_rate_input.value,
            market_rate_input.value,
            maturity_input.value,
            frequency_input.value
        )
        print(f"Bond Price: {price}")

calculate_button = widgets.Button(description="Calculate Price")
calculate_button.on_click(on_calculate_button_clicked)

# Display all widgets
display(face_value_input, coupon_rate_input, market_rate_input, maturity_input, frequency_input, calculate_button, output)
```

### Bond Valuation: Remaining Time to Maturity and Bond Price

* The closer to maturity, the closer the price of the bond is to its face value.
    * Fewer coupon payments as the bond gets closer to maturity.
    * If the coupon rate is higher than the market rate, the price of the bond decreases as it approaches maturity while the price of the bond increases as it approaches maturity if the coupon rate is lower than the market rate.

```python
#!pip install ipywidgets matplotlib scipy

import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d
import ipywidgets as widgets
from IPython.display import display, clear_output

# ----------------------------
# ORIGINAL DISCRETE LOGIC
# ----------------------------
def bond_price(remaining_maturity, coupon_rate, face_value, ytm, coupon_freq):
    """
    Computes bond price using discrete period logic.
    """
    periods = int(np.ceil(remaining_maturity * coupon_freq))
    if periods == 0:
        return face_value  # At maturity
    
    period_yield = ytm / coupon_freq
    coupon_pmt = (face_value * coupon_rate) / coupon_freq
    
    # Present value of coupon payments (annuity)
    pv_coupons = coupon_pmt * (1 - (1 + period_yield)**(-periods)) / period_yield
    # Present value of face value
    pv_face = face_value / ((1 + period_yield)**periods)
    
    return pv_coupons + pv_face

def plot_bond_price_dynamics(face_value, ytm_percent, coupon1_percent, coupon2_percent, coupon_freq):
    """
    Recreates the discrete approach with interpolation for a smooth curve.
    """
    # Convert percentages to decimals
    ytm = ytm_percent / 100
    coupon1 = coupon1_percent / 100
    coupon2 = coupon2_percent / 100
    
    # Discrete maturities array, from 5 years down to 0
    maturities = np.array([5.0, 4.5, 4.0, 3.5, 3.0, 2.5, 2.0, 1.5, 1.0, 0.5, 0.0])
    
    # Compute discrete prices for each coupon rate
    bond_prices_1 = [bond_price(t, coupon1, face_value, ytm, coupon_freq) for t in maturities]
    bond_prices_2 = [bond_price(t, coupon2, face_value, ytm, coupon_freq) for t in maturities]
    
    # Interpolate for a smooth curve (quadratic)
    sorted_idx = np.argsort(maturities)
    continuous_maturity = np.linspace(5, 0, 500)  # from 5 → 0
    interp_func_1 = interp1d(maturities[sorted_idx], np.array(bond_prices_1)[sorted_idx],
                             kind='quadratic', fill_value='extrapolate')
    interp_func_2 = interp1d(maturities[sorted_idx], np.array(bond_prices_2)[sorted_idx],
                             kind='quadratic', fill_value='extrapolate')
    
    # Plotting the graph
    plt.figure(figsize=(9, 6), dpi=100)
    plt.plot(continuous_maturity, interp_func_1(continuous_maturity),
             label=f'{coupon1_percent:.0f}% Coupon Bond (YTM {ytm_percent:.0f}%)')
    plt.plot(continuous_maturity, interp_func_2(continuous_maturity),
             label=f'{coupon2_percent:.0f}% Coupon Bond (YTM {ytm_percent:.0f}%)')
    
    # Discrete data points
    plt.scatter(maturities, bond_prices_1, s=80)
    plt.scatter(maturities, bond_prices_2, s=80)
    
    # Flip axis: 5 years on left to 0 years on right
    plt.gca().invert_xaxis()
    plt.title("Bond Price Dynamics (Discrete Period Logic)")
    plt.xlabel("Remaining Maturity (Years)")
    plt.ylabel("Bond Price ($)")
    plt.grid(True, alpha=0.3, linestyle=':')
    plt.legend()
    plt.show()

# -------------------------
# WIDGETS FOR INTERACTIVITY
# -------------------------
face_value_widget = widgets.FloatText(
    value=10000,
    description='Face Value:'
)
ytm_widget = widgets.FloatText(
    value=10.0,  # 8% Yield to Maturity default (as per new default)
    description='YTM (%):'
)
coupon1_widget = widgets.FloatText(
    value=8.0,
    description='Coupon1 (%):'
)
coupon2_widget = widgets.FloatText(
    value=12.0,
    description='Coupon2 (%):'
)
# Removed option 1; now only semi-annual (2) or quarterly (4)
coupon_freq_widget = widgets.Dropdown(
    options=[2, 4, 12],
    value=2,
    description='Payments/year:'
)

output = widgets.Output()

def on_recalculate_clicked(_):
    with output:
        clear_output()
        plot_bond_price_dynamics(
            face_value_widget.value,
            ytm_widget.value,
            coupon1_widget.value,
            coupon2_widget.value,
            coupon_freq_widget.value
        )

recalculate_button = widgets.Button(description="Recalculate Graph")
recalculate_button.on_click(on_recalculate_clicked)

# Display widgets and button, then immediately generate the graph with default values.
display(
    face_value_widget,
    ytm_widget,
    coupon1_widget,
    coupon2_widget,
    coupon_freq_widget,
    recalculate_button,
    output
)

# Generate the default graph upon first execution.
on_recalculate_clicked(None)
```

### Bond Valuation: Yield Measures for Fixed-Income Securities
* For the plain fixed-income securities that do not have embedded options current yield and the yield to maturity are the two conventional yield measures.
* Current yield
    * Relates the annual coupon interest to the market price.
    * It is a simple but insufficient measure of return to calculate.
    $$ Current \ Yield = \frac{Annual \ Coupon \ Payment}{Market \ Price} $$
    * The current yield has some significant shortcomings:
        * It only takes coupon interest into account and ignores capital gains/losses.
            * It does not take into account the capital gain that an investor will receive from purchasing a discount-traded bond and holding it to maturity.
            * If a bond is purchased at premium and held until maturity, the capital loss that the investor will face is also not taken into account.
        * The time value of money is ignored.
---

### Bond Valuation: Yield Measures for Fixed-Income Securities
**What is the current yield of a bond at the issuance that pays annual coupons of €800, has 10 years remaining to maturity when the market interest rate is 9%? The nominal value of the bond is €10,000.**

$ P = \left( \sum_{t=1}^{10} \frac{800}{(1 + 0.09)^{t}} \right) + \frac{10{,}000}{(1 + 0.09)^{10}} = 9{,}358.23 \qquad \rightarrow \qquad Current \ Yield = \frac{Annual \ Coupon \ Payment}{Market \ Price} = \frac{800}{9,358.23} = 8.55\% $

**What if the market interest rate is at 6%?**

$ P = \left( \sum_{t=1}^{10} \frac{800}{(1+0.06)^t} \right) + \frac{10,000}{(1+0.06)^{10}} = 11,472.02 \qquad \rightarrow \qquad Current \ Yield = \frac{800}{11,472.02} = 6.97\% $

**What if the bond pays semiannual coupons of €400 and the market interest rate is at 7%?**

$ P = \left( \sum_{t=1}^{20} \frac{400}{(1+0.035)^t} \right) + \frac{10,000}{(1+0.035)^{20}} = 10,710.62 \qquad \rightarrow \qquad Current \ Yield = \frac{800}{10,710.62} = 7.47\% $

### Bond Valuation: Yield-to-maturity (YTM)
* Given the shortcomings of the current yield, a more appropriate and widely used measure is yield-to-maturity (YTM) which is the interest rate that, if the bond is held to maturity, will make the present value of the remaining cash flows equal to the sum of its price and any accumulated interest.
    * To calculate the YTM, we need to solve the following bond pricing equation for r.
    * The YTM is akin to the internal rate of return (IRR).
    * Since calculating the YTM requires iteration, as is the case for IRR, either spreadsheet solutions or financial calculators may be used to obtain the YTM of a bond.
    * We solve the bond pricing formula for r;
    $$ P = \left( \sum_{t=1}^{n} \frac{C}{(1+r)^t} \right) + \frac{FV}{(1+r)^n} $$

```python
#!pip install ipywidgets scipy

import numpy as np
from scipy.optimize import brentq
import ipywidgets as widgets
from IPython.display import display, clear_output

# ---------------------------------------------------------
# Utility Functions
# ---------------------------------------------------------
def npv(rate, cash_flows):
    """Calculate net present value for given cash flows discounted at 'rate'."""
    return sum(cf / (1 + rate) ** t for t, cf in enumerate(cash_flows))

def calculate_ytm(face_value, current_price, annual_coupon, years, payments_per_year):
    """
    Calculate the yield to maturity (YTM) based on cash flows.
    
    Parameters:
      face_value         : Nominal value to be received at maturity.
      current_price      : Current market price of the bond.
      annual_coupon      : Coupon amount for the entire year.
      years              : Time to maturity in years (float).
      payments_per_year  : Number of coupon payments per year (int).
      
    Returns:
      yield_per_period   : The yield per coupon payment period (decimal).
      annual_yield       : The nominal annualized yield (decimal).
      cash_flows         : Generated list of cash flows over each period.
    """
    periods = int(years * payments_per_year)
    coupon_per_period = annual_coupon / payments_per_year
    
    # t = 0: negative (purchase); t = 1..(periods-1): coupon_per_period;
    # t = periods: coupon_per_period + face_value.
    cash_flows = [-current_price] + [coupon_per_period] * (periods - 1) + [coupon_per_period + face_value]
    
    # Solve for the discount rate that makes NPV = 0
    yield_per_period = brentq(lambda r: npv(r, cash_flows), 0.0001, 1.0)
    annual_yield = yield_per_period * payments_per_year
    return yield_per_period, annual_yield, cash_flows

# ---------------------------------------------------------
# Widget Event Handler
# ---------------------------------------------------------
def recalc_yield(_):
    with output:
        clear_output()
        face_val = face_value_widget.value
        curr_price = current_price_widget.value
        annual_coupon = coupon_widget.value
        years = maturity_widget.value
        freq = frequency_widget.value
        
        ytm_per_period, annual_ytm, cf = calculate_ytm(face_val, curr_price, annual_coupon, years, freq)
        
        print("Cash Flows:")
        for t, amount in enumerate(cf):
            print(f"t = {t}: {amount:,.2f}")
            
        print("\nCalculated Yield to Maturity:")
        print(f"Yield per period: {ytm_per_period:.4%}")
        print(f"Annualized Nominal Yield: {annual_ytm:.4%}")

# ---------------------------------------------------------
# Widget Definitions with Layout/Style
# ---------------------------------------------------------
label_style = {'description_width': '150px'}
input_layout = widgets.Layout(width='330px')

face_value_widget = widgets.FloatText(
    value=10000.00,
    step=0.01,
    description='Face Value:',
    style=label_style,
    layout=input_layout
)
current_price_widget = widgets.FloatText(
    value=9000.00,
    step=0.01,
    description='Current Price:',
    style=label_style,
    layout=input_layout
)
coupon_widget = widgets.FloatText(
    value=1200,
    step=50,
    description='Annual Coupon:',
    style=label_style,
    layout=input_layout
)
maturity_widget = widgets.FloatText(
    value=10,
    step=0.5,
    description='Maturity (Years):',
    style=label_style,
    layout=input_layout
)
frequency_widget = widgets.Dropdown(
    options=[1, 2, 4, 12],
    value=1,
    description='Payments/Year:',
    style=label_style,
    layout=input_layout
)

recalc_button = widgets.Button(
    description="Recalculate YTM",
    layout=widgets.Layout(width='150px')
)

output = widgets.Output()
recalc_button.on_click(recalc_yield)

# ---------------------------------------------------------
# Display the Interactive Interface
# ---------------------------------------------------------
display(face_value_widget,
        current_price_widget,
        coupon_widget,
        maturity_widget,
        frequency_widget,
        recalc_button,
        output)

# Generate the default output upon first run.
recalc_yield(None)
```

### Bond Valuation
**Why is the YTM is superior to the current yield?**
* It takes not only the current coupon income into account but also the capital gain or loss the investor will face by holding the bond to maturity.
* The YTM also takes the timing of cash flows into account in contrast to the current yield.

**Reinvestment risk and the YTM**

* An important assumption of the YTM is that the bond's coupon payments are reinvested, and the reinvestment rate of the coupon payments equals the YTM.
* As the reinvestment rates differ from the initial YTM, the realized rate of return will also differ from the initial YTM.
* This is called the reinvestment risk and may become a major source of risk in volatile markets.

---

### The Term Structure Of Interest Rates

* The relationship between the yields on comparable securities with different maturities is the term structure of interest rates.

* The yield curve is a graphical representation of the YTM as a function of time to maturity.
* The yield curve is one of the main elements of bond valuation and a key indicator for fixed-income market investors.
* It allows investors to compare their expectations of future interest rates with market expectations.
* The returns on the yield curve are not calculated directly using the returns observed in the market.
    * Assuming that the securities are issued with a zero coupon, the theoretical interest rates to be paid are calculated.
    * Thus, the rates of return on the yield curve can be used as spot rates corresponding to a certain maturity.
* It is useful in calculating the present value of a cash flow to be received in the future.

```python
#!pip install fredapi
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from fredapi import Fred
from IPython.display import HTML

fred = Fred(api_key=os.environ['FRED_API_KEY'])

# Treasury maturities and FRED codes
maturities = {
    "1 Mo": "DGS1MO",
    "3 Mo": "DGS3MO",
    "6 Mo": "DGS6MO",
    "1 Yr": "DGS1",
    "2 Yr": "DGS2",
    "3 Yr": "DGS3",
    "5 Yr": "DGS5",
    "7 Yr": "DGS7",
    "10 Yr": "DGS10",
    "20 Yr": "DGS20",
    "30 Yr": "DGS30"
}
maturity_labels = list(maturities.keys())
maturity_positions = np.arange(len(maturity_labels))

# Dates from 1985 to 2024
date_range = pd.date_range(start='1985-12-31', end='2024-12-31', freq='A')

# Fetch data once using fredapi
print("Fetching data from FRED...")
start_date = (date_range.min() - pd.Timedelta(days=10)).strftime('%Y-%m-%d')

all_series = {}
for label, code in maturities.items():
    print(f"  {label}...", end=" ")
    all_series[code] = fred.get_series(code, observation_start=start_date)
    print("ok")

yield_data = pd.DataFrame(all_series).ffill()

# Prepare yield curves data
yield_curves = {}
for date in date_range:
    closest_date = yield_data.index.asof(date)
    yields = yield_data.loc[closest_date].values
    yield_curves[date.strftime('%Y-%m-%d')] = yields

# Set up figure and axes
fig, ax = plt.subplots(figsize=(10, 6))
plt.subplots_adjust(left=0.1, right=0.9, top=0.9, bottom=0.2)

yield_curve_line, = ax.plot([], [], marker='o', linestyle='-', color='navy')

ax.set_title("U.S. Treasury Yield Curve (Annual from 1985 to 2024)", fontsize=14, pad=15)
ax.set_xlabel("Maturity", fontsize=12)
ax.set_ylabel("Yield (%)", fontsize=12)
ax.set_xlim(0, len(maturity_labels)-1)
ax.set_xticks(maturity_positions)
ax.set_xticklabels(maturity_labels)
ax.grid(True)

date_text = ax.text(0.5, 0.95, "", transform=ax.transAxes, fontsize=12, ha="center", va="top",
                    bbox=dict(facecolor='white', alpha=0.8, boxstyle='round,pad=0.4'))

def update(frame):
    date = list(yield_curves.keys())[frame]
    yields = yield_curves[date]
    yield_curve_line.set_data(maturity_positions, yields)
    ax.set_ylim(0, np.nanmax(yield_data.values) + 2)
    date_text.set_text(f"Yield Curve as of: {date}")
    return yield_curve_line, date_text

plt.close(fig)

ani = animation.FuncAnimation(
    fig, update, frames=len(yield_curves),
    interval=500, blit=False, repeat=True
)

HTML(ani.to_jshtml())
```

### The Term Structure of Interest Rates

**Calculating the first discount factor**

$P_1 = CF_1 \times DF_{0,1} \quad \Rightarrow \quad 100 = 110 \times DF_{0,1} \quad \Rightarrow \quad DF_{0,1} = 0.9091$

**Calculating the second discount factor**

$P_2 = CF_1 \times DF_{0,1} + CF_2 \times DF_{0,2}  \quad \Rightarrow \quad 90 = 5 \times 0.9091 + 105 \times DF_{0,2} \quad \Rightarrow \quad DF_{0,2} = 0.8139$

**Computing the forward rate using the interest rates**

$r_{0,1} = \frac{1}{DF_{0,1}} - 1 = \frac{1}{0.9091} - 1 = 10\% \qquad \text{and} \qquad r_{0,2} = \sqrt{\frac{1}{DF_{0,2}}} - 1 = \sqrt{\frac{1}{0.8139}} - 1 = 10.84\%$

$ (1+r_{0,2})^2 = (1+r_{0,1}) \times (1+r_{1,2}) \qquad \rightarrow \qquad \frac {(1+10.84\%)^2} {(1+10\%)} = (1+r_{1,2}) \qquad \rightarrow \qquad r_{1,2}=11.69\%$

**Computing the forward rate using the discount factors**

$DF_{0,1} = (1 + r_{0,1})^{-1} \quad \text{and} \quad DF_{0,2} = (1 + r_{0,2})^{-2}$

$DF_{0,2} = DF_{0,1} \times DF_{1,2} \quad \Rightarrow \quad \frac{DF_{0,2}}{DF_{0,1}} = DF_{1,2} = (1 + r_{1,2})^{-1} \quad \Rightarrow \quad \frac{0.8139}{0.9091} = (1 + r_{1,2})^{-1} \quad \Rightarrow \quad r_{1,2} = 11.69\%$

### Default Risk
* The cash flows that a purchaser of a bond with credit risk expects to receive may be less than the promised amount.
    * As a result, investors pay less for bonds with credit risk than they would for an otherwise identical default-free bond.
    * Because the yield to maturity for a bond is calculated using the promised cash flows, the yield of bonds with credit risk will be higher than that of otherwise identical default-free bonds.
* Suppose that the one-year, zero-coupon Treasury bill has a yield to maturity of 4%. What are the price and yield of a one-year, $1000, zero-coupon bond issued by AGE Corporation, which has zero probability of default?

    * $P=1000/(1+YTM)=1000/(1.04)=\$961.54$
