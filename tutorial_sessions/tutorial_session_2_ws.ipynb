### Question 1.1

Alpha Industrial Solutions is deciding whether to replace its aging production assembly line now or in five years. 
-   The current assembly line has a market salvage value of $\$4.0$ million and a tax book value of $\$2.5$ million. 
-   If the company continues using the old line, it will incur annual maintenance expenses of $\$1.2$ million at the end of each year for the next five years. 
-   Depreciation on the old line is $\$500,000$ per year for the next five years, at which point its book value will be $\$0$. At the end of five years, it is estimated to have a salvage value of $\$400,000$.

A state-of-the-art replacement system costs $\$6.5$ million today. It is expected to require maintenance costs of $\$600,000$ at the end of each year during its five-year economic life. 
-   At the end of five years, the new system will have a market salvage value of $\$1.5$ million. It will be fully depreciated using the straight-line method over five years to a zero book value. 
-   In five years, a replacement system will cost $\$6.0$ million regardless of the choice made today. 
-   The corporate tax rate is $25\%$ and the appropriate discount rate is $8\%$. The company generates enough revenue to fully utilize all depreciation tax shields.

Should the company replace the old assembly line today or wait until the end of five years?

---

### Question 1.1: Solution

To make the decision, we calculate the Net Present Value (NPV) of the incremental cash flows from replacing the system today. The cost of a machine five years from now is irrelevant as it is a common cost to both alternatives.

**Initial Investment (Year 0 Incremental Cash Flow):**

Outlay for new system: $-\$6,500,000$

After-tax salvage value of old system:
   Salvage value: $+\$4,000,000$
   
   Tax on gain: $-(\$4,000,000 - \$2,500,000) \times 0.25 = -\$375,000$
   
   Net proceeds: $\$4,000,000 - \$375,000 = +\$3,625,000$

Total Year 0 Cash Flow ($CF_0$): $-\$6,500,000 + \$3,625,000 = -\$2,875,000$

**Incremental Operating Cash Flow (Years 1-5):**

- Savings in maintenance costs: $\$1,200,000 - \$600,000 = \$600,000$
   
   After-tax savings: $\$600,000 \times (1 - 0.25) = +\$450,000$

- Change in depreciation tax shield:
   
   New depreciation: $\$6,500,000 / 5 = \$1,300,000$
   
   Old depreciation: $\$500,000$
   
   Increase in depreciation: $\$1,300,000 - \$500,000 = \$800,000$
   
   Additional tax shield: $\$800,000 \times 0.25 = +\$200,000$

- Total Incremental OCF (Years 1-5): $\$450,000 + \$200,000 = \$650,000$

**Terminal Incremental Cash Flow (Year 5):**

- After-tax salvage value of new system:
   
   Salvage value: $+\$1,500,000$
   
   Tax on gain: $-(\$1,500,000 - \$0) \times 0.25 = -\$375,000$
   
   Net proceeds: $+\$1,125,000$

- Forgone after-tax salvage value of old system:
   
   Salvage value: $-\$400,000$
   
   Tax on gain: $+(\$400,000 - \$0) \times 0.25 = +\$100,000$
   
   Net forgone proceeds: $-\$ 300,000$

- Total Terminal Cash Flow: $\$ 1,125,000 - \$300,000 = +\$825,000$

**NPV Calculation:**

The total incremental cash flow in year 5 is the operating cash flow plus the net terminal cash flow: $\$650,000 + \$825,000 = \$1,475,000$.

$$ NPV = CF_0 + \sum_{t=1}^{4} \frac{CF_t}{(1+r)^t} + \frac{CF_5}{(1+r)^5} $$

$$ NPV = -\$2,875,000 + \frac{\$650,000}{1.08^1} + \frac{\$650,000}{1.08^2} + \frac{\$650,000}{1.08^3} + \frac{\$650,000}{1.08^4} + \frac{\$1,475,000}{1.08^5} $$

$$ NPV = \$281,741.36 $$

**Since the NPV of the incremental cash flows is positive ($\$281,741.36$), Alpha Industrial Solutions should replace the old assembly line today.**

### Question 1.2

**Gamma Systems** is considering a risk-free financial security that promises the following cash flows: $\$200$ in one year, $\$250$ in two years, and $\$300$ in three years. Assume the current risk-free interest rate is 5% per annum.

a) Calculate the theoretical "no-arbitrage" price of this security today.

b) If the security is currently trading in the market for $\$710$, identify the arbitrage opportunity and describe the strategy to capture a risk-free profit.

---

### Question 1.2: Solution

a) The no-arbitrage price is the present value (PV) of the guaranteed future cash flows, discounted at the risk-free rate.

$$ PV = \frac{C_1}{(1+r)^1} + \frac{C_2}{(1+r)^2} + \frac{C_3}{(1+r)^3} $$

$$ PV = \frac{\$200}{(1.05)^1} + \frac{\$250}{(1.05)^2} + \frac{\$300}{(1.05)^3} $$

$$ PV = \$190.4762 + \$226.7574 + \$259.1513 \quad \Rightarrow \quad PV = \$676.3849 $$

The no-arbitrage price of the security is **$\$676.3849$**.

b) The security's fair value is $\$676.3849$, but it is trading at $\$710$. Since the market price is higher than the theoretical value, the security is overpriced.

**Arbitrage Cash Flow Table:**

| Transaction | Cash Flow Today ($t=0$) | Cash Flow in 1 Year ($t=1$) | Cash Flow in 2 Years ($t=2$) | Cash Flow in 3 Years ($t=3$) |
| :--- | :---: | :---: | :---: | :---: |
| Sell Overpriced Security | $+\$710.0000$ | $-\$200.00$ | $-\$250.00$ | $-\$300.00$ |
| Invest in Risk-Free Rate | $-\$676.3849$ | $+\$200.00$ | $+\$250.00$ | $+\$300.00$ |
| **Net Cash Flow** | **$+\$33.6151$** | **$\$0.00$** | **$\$0.00$** | **$\$0.00$** |

### Question 1.3

You have the following spot yield curve information for risk-free government securities:

| Maturity | Spot Rate ($z_t$) |
| :--- | ---: |
| 1 year | 4.0% |
| 2 years | 4.2% |
| 4 years | 4.8% |
| 5 years | 5.0% |

a) Calculate the current prices of zero-coupon bonds with a $1,000 face value and maturities of 1, 2, 4, and 5 years.

b) Determine the one-year forward rates starting at $t = 1$ and $t = 4$.

c) A bond issued by Delta Dynamics with a $20,000 face value, a 5-year maturity, and 7% annual coupon payments is currently priced at $21,790.49. What is the implied 3-year spot rate ($z_3$)? 

*(Assume that this bond's cash flows share the same risk as the assets used to derive the yield curve above, and that the next coupon payment occurs in exactly one year.)*

---

### Question 1.3: Solution

a) The price of a zero-coupon bond is its face value discounted by the corresponding spot rate.

$$P_t = \frac{FV}{(1+z_t)^t}$$

$P_1 = \frac{\$1,000}{(1.04)^1} = \$961.5385$

$P_2 = \frac{\$1,000}{(1.042)^2} = \$921.0227$

$P_4 = \frac{\$1,000}{(1.048)^4} = \$828.9882$

$P_5 = \frac{\$1,000}{(1.05)^5} = \$783.5262$

b) The one-year forward rate is calculated using the spot rates: $(1+z_t)^t(1+f_{t, t+1}) = (1+z_{t+1})^{t+1}$.

For $f_{1,2}$:
$$1+f_{1,2} = \frac{(1.042)^2}{(1.04)^1} = \frac{1.085764}{1.04} \approx 1.043996 \quad \Rightarrow \quad f_{1,2} \approx 4.40\%$$

For $f_{4,5}$:
$$1+f_{4,5} = \frac{(1.05)^5}{(1.048)^4} = \frac{1.276282}{1.206290} \approx 1.058023 \quad \Rightarrow \quad f_{4,5} \approx 5.80\%$$

c) The bond price is the sum of discounted cash flows. Annual coupon $C = \$20,000 \times 0.07 = \$1,400$.

$$PV(C_3) = Price - \left[ \frac{C_1}{(1+z_1)^1} + \frac{C_2}{(1+z_2)^2} + \frac{C_4}{(1+z_4)^4} + \frac{C_5+FV}{(1+z_5)^5} \right]$$

$$PV(C_3) = \$21,790.49 - \left[ \frac{1,400}{1.04^1} + \frac{1,400}{1.042^2} + \frac{1,400}{1.048^4} + \frac{21,400}{1.05^5} \right]$$

$$PV(C_3) = \$21,790.49 - [1,346.1538 + 1,289.4318 + 1,160.5835 + 16,767.4608] \quad \Rightarrow \quad PV(C_3) = \$1,226.8601$$

$$\frac{1,400}{(1+z_3)^3} = \$1,226.8601 \quad \Rightarrow \quad (1+z_3)^3 = \frac{1,400}{1,226.8601} \approx 1.1411245$$

$$1+z_3 = (1.1411245)^{1/3} \approx 1.045012 \quad \Rightarrow \quad z_3 \approx 4.50\%$$

### Question 1.4

Eta Dynamics has just paid an annual dividend of $3.00 per share. 

* Market analysts are projecting that the annual dividend will grow by $0.25 per year over each of the next five years. 

* Starting from Year 6, the company’s earnings are expected to grow at a constant rate of 5.0% per year, and Eta Dynamics intends to maintain a constant dividend payout ratio indefinitely.

If the equity cost of capital for Eta Dynamics is 11.0% per year, what is the current value of one share of stock according to the dividend-discount model?

---

### Question 1.4: Solution

$D_1 = 3.00 + 0.25 = 3.25$
$D_2 = 3.25 + 0.25 = 3.50$
$D_3 = 3.50 + 0.25 = 3.75$
$D_4 = 3.75 + 0.25 = 4.00$
$D_5 = 4.00 + 0.25 = 4.25$

$$ D_6 = D_5 \times (1 + g) = 4.25 \times 1.05 = 4.4625 $$

The terminal price represents the present value at $t=5$ of all dividends from $t=6$ to infinity:
$$ P_5 = \frac{D_6}{r_E - g} = \frac{4.4625}{0.11 - 0.05} = \frac{4.4625}{0.06} = 74.3750 $$

$$ P_0 = \frac{D_1}{(1+r_E)^1} + \frac{D_2}{(1+r_E)^2} + \frac{D_3}{(1+r_E)^3} + \frac{D_4}{(1+r_E)^4} + \frac{D_5 + P_5}{(1+r_E)^5} $$

$$ P_0 = \frac{3.25}{1.11^1} + \frac{3.50}{1.11^2} + \frac{3.75}{1.11^3} + \frac{4.00}{1.11^4} + \frac{4.25 + 74.375}{1.11^5} \quad \Rightarrow \quad P_0 = 57.8056 $$

### Question 1.5

Lambda Systems is a privately held developer of high-precision monitoring software based in Munich. As part of your strategic expansion, in late 2024 you initiate discussions with Lambda's founder about acquiring the business at the end of 2024. Estimate the value of Lambda Systems per share using a discounted Free Cash Flow (FCF) approach and the following data:

-   Debt: $50 million

-   Excess cash: $150 million

-   Shares outstanding: 80 million

-   Expected FCF in 2025: $70 million

-   Expected FCF in 2026: $80 million

-   Future FCF growth rate beyond 2026: 4%

-   Weighted-average cost of capital (WACC): 9.0%

---

### Question 1.5: Solution

The valuation is performed as of the end of 2024 ($t=0$).

$$ FCF_{2027} = FCF_{2026} \times (1 + g) = \$80,000,000 \times (1.04) = \$83,200,000 $$

$$ V_{2026} = \frac{FCF_{2027}}{WACC - g} = \frac{\$83,200,000}{0.09 - 0.04} = \$1,664,000,000 $$

$$ V_0 = \frac{FCF_{2025}}{(1+WACC)^1} + \frac{FCF_{2026} + V_{2026}}{(1+WACC)^2} $$
$$ V_0 = \frac{\$70,000,000}{(1.09)^1} + \frac{\$80,000,000 + \$1,664,000,000}{(1.09)^2} \quad \Rightarrow \quad V_0 = \$1,532,110,091.75 $$

$$ \text{Equity Value} = V_0 - \text{Debt} + \text{Cash} $$
$$ \text{Equity Value} = \$1,532,110,091.75 - \$50,000,000 + \$150,000,000 = \$1,632,110,091.75 $$

$$ \text{Price per Share} = \frac{\text{Equity Value}}{\text{Shares Outstanding}} $$
$$ \text{Price per Share} = \frac{\$1,632,110,091.75}{80,000,000} \approx \$20.4014 $$

### Question 2.1

Consider the following information on two stocks:

| State of Economy | Probability of State | Rate of Return: Stock Delta | Rate of Return: Stock Epsilon |
| :--- | :---: | :---: | :---: |
| Bull Market | 0.60 | 10% | 2% |
| Bear Market | 0.40 | 6% | 14% |

a) Calculate the expected return and variance for each stock.

b) Calculate the covariance between Stock Delta and Stock Epsilon.

c) Derive the expression for the weight $w_1$ that minimizes the portfolio variance and calculate the weight for Stock Delta in the MVP.

---

### Question 2.1: Solution

**a) Expected Return and Variance**

$E(R_D) = 0.60(0.10) + 0.40(0.06) = 0.084$ (or 8.4%)

$\sigma_D^2 = 0.60(0.10 - 0.084)^2 + 0.40(0.06 - 0.084)^2 \quad \Rightarrow \quad \sigma_D^2 = 0.60(0.000256) + 0.40(0.000576) = 0.000384$

$E(R_E) = 0.60(0.02) + 0.40(0.14) = 0.068$ (or 6.8%)

$\sigma_E^2 = 0.60(0.02 - 0.068)^2 + 0.40(0.14 - 0.068)^2 \quad \Rightarrow \quad \sigma_E^2 = 0.60(0.002304) + 0.40(0.005184) = 0.003456$

**b) Covariance**

$\text{Cov}(R_D, R_E) = \sum P_i [R_{D,i} - E(R_D)][R_{E,i} - E(R_E)]$

$\text{Cov}(R_D, R_E) = 0.60(0.10 - 0.084)(0.02 - 0.068) + 0.40(0.06 - 0.084)(0.14 - 0.068) \quad \Rightarrow \quad \text{Cov}(R_D, R_E) = -0.0004608 - 0.0006912 = -0.001152$

**c) Minimum Variance Portfolio (MVP)**

Substitute $w_2 = 1 - w_1$ into the variance formula:
$\sigma_p^2 = w_1^2 \sigma_1^2 + (1-w_1)^2 \sigma_2^2 + 2w_1(1-w_1)\text{Cov}(R_1, R_2)$

Take the derivative with respect to $w_1$ and set it to zero:
$\frac{d\sigma_p^2}{dw_1} = 2w_1\sigma_1^2 - 2(1-w_1)\sigma_2^2 + 2(1-2w_1)\text{Cov}(R_1, R_2) = 0$

$w_1\sigma_1^2 - \sigma_2^2 + w_1\sigma_2^2 + \text{Cov}(R_1, R_2) - 2w_1\text{Cov}(R_1, R_2) = 0$

$w_1(\sigma_1^2 + \sigma_2^2 - 2\text{Cov}(R_1, R_2)) = \sigma_2^2 - \text{Cov}(R_1, R_2)$

$$w_1 = \frac{\sigma_2^2 - \text{Cov}(R_1, R_2)}{\sigma_1^2 + \sigma_2^2 - 2\text{Cov}(R_1, R_2)}$$

$w_D = \frac{0.003456 - (-0.001152)}{0.000384 + 0.003456 - 2(-0.001152)} \quad \Rightarrow \quad w_D = 0.75$

### Question 2.2

Using the data in the following table for the returns of Stock Delta and Stock Epsilon:

| Year | 2018 | 2019 | 2020 | 2021 | 2022 | 2023 |
| :--- | :--- | :--- | :--- | :--- | :--- | :--- |
| Stock Delta | 15% | –8% | 22% | 4% | –12% | 18% |
| Stock Epsilon | 20% | 5% | –10% | 28% | –4% | 15% |

a. Estimate the average return and volatility for each stock.

b. Estimate the covariance between the stocks.

c. Estimate the correlation between these two stocks.

d. Consider a portfolio that maintains a 50% weight on stock Delta and a 50% weight on stock Epsilon. What is the return each year of this portfolio? Compute the average return and volatility of the equally-weighted portfolio.

e. Calculate the volatility of a portfolio that is 60% invested in stock Delta and 40% invested in stock Epsilon.

---

### Question 2.2: Solution

**a. Average Return and Volatility**

$ \bar{R}_{Delta} = \frac{0.15 - 0.08 + 0.22 + 0.04 - 0.12 + 0.18}{6} = 0.0650 $

$ \text{Var}(R_{Delta}) = \frac{1}{5} [ (0.15-0.0650)^2 + ... + (0.18-0.0650)^2 ] = 0.0201 \quad \Rightarrow \quad \text{SD}(R_{Delta}) = \sqrt{0.0201} = 0.1417 $

$ \bar{R}_{Epsilon} = \frac{0.20 + 0.05 - 0.10 + 0.28 - 0.04 + 0.15}{6} = 0.0900 $

$ \text{Var}(R_{Epsilon}) = \frac{1}{5} [ (0.20-0.0900)^2 + ... + (0.15-0.0900)^2 ] = 0.0213 \quad \Rightarrow \quad \text{SD}(R_{Epsilon}) = \sqrt{0.0213} = 0.1459 $

**b. Covariance**

$ \text{Cov}(R_{Delta}, R_{Epsilon}) = \frac{1}{5} [ (0.15-0.0650)(0.20-0.0900) + ... + (0.18-0.0650)(0.15-0.0900) ] \quad \Rightarrow \quad \text{Cov}(R_{Delta}, R_{Epsilon}) = 0.0024 $

**c. Correlation**

$ \text{Corr}(R_{Delta}, R_{Epsilon}) = \frac{\text{Cov}(R_{Delta}, R_{Epsilon})}{\text{SD}(R_{Delta})\text{SD}(R_{Epsilon})} = \frac{0.0024}{(0.1417)(0.1459)} = 0.1160 $

**d. Equally-Weighted Portfolio**

The annual returns are: $0.1750$, $-0.0150$, $0.0600$, $0.1600$, $-0.0800$, $0.1650$.

$ \bar{R}_p = \frac{0.1750 - 0.0150 + 0.0600 + 0.1600 - 0.0800 + 0.1650}{6} = 0.0775 $

$ \text{Var}(R_p) = \frac{1}{5} [ (0.1750 - 0.0775)^2 + ... + (0.1650 - 0.0775)^2 ] = 0.0115 \quad \Rightarrow \quad \text{SD}(R_p) = \sqrt{0.0115} = 0.1072 $

**e. Volatility of 60/40 Portfolio**

$ \text{Var}(R_p) = w_{Delta}^2\text{Var}(R_{Delta}) + w_{Epsilon}^2\text{Var}(R_{Epsilon}) + 2w_{Delta}w_{Epsilon}\text{Cov}(R_{Delta}, R_{Epsilon})$

$\Rightarrow \quad \text{Var}(R_p) = (0.60)^2(0.0201) + (0.40)^2(0.0213) + 2(0.60)(0.40)(0.0024) \quad \Rightarrow \quad \text{Var}(R_p) = 0.0118 $

$ \text{SD}(R_p) = \sqrt{0.0118} = 0.1086 $

### Question 2.3

Suppose Omega Corporation and Sigma Corporation have the expected returns and volatilities shown below, with a correlation of 20%.

| Asset | Expected Return | Standard Deviation |
| :--- | :--- | :--- |
| Omega Corporation | 10% | 20% |
| Sigma Corporation | 15% | 30% |

a. Calculate the expected return and the volatility of a portfolio that is equally invested in Omega's and Sigma's stock.

b. Calculate the expected return and the volatility of a portfolio that consists of a long position of $25,000 in Omega and a short position of $5,000 in Sigma.

c. Determine the portfolio weights ($w_{Omega}$ and $w_{Sigma}$) required to achieve a target portfolio volatility (standard deviation) of exactly 21%.

---

### Question 2.3: Solution

**a. For an equally weighted portfolio, $w_{Omega} = 0.50$ and $w_{Sigma} = 0.50$.**

$E(R_p) = (0.50)(0.10) + (0.50)(0.15) = 0.1250$ or 12.50%

$\sigma_p^2 = w_{Omega}^2\sigma_{Omega}^2 + w_{Sigma}^2\sigma_{Sigma}^2 + 2w_{Omega}w_{Sigma}\sigma_{Omega}\sigma_{Sigma}\rho_{Omega,Sigma}$

$\sigma_p^2 = (0.50)^2(0.20)^2 + (0.50)^2(0.30)^2 + 2(0.50)(0.50)(0.20)(0.30)(0.20) \quad \Rightarrow \quad \sigma_p^2 = 0.0385 \quad \Rightarrow \quad \sigma_p = \sqrt{0.0385} \approx 0.1962$ or 19.62%

---

**b. Total value of the portfolio is $25,000 - $5,000 = $20,000.**

$w_{Omega} = \frac{\$25,000}{\$20,000} = 1.25 \qquad \text{and} \qquad w_{Sigma} = \frac{-\$5,000}{\$20,000} = -0.25$

$E(R_p) = (1.25)(0.10) + (-0.25)(0.15) = 0.1250 - 0.0375 = 0.0875$ or 8.75%

$\sigma_p^2 = (1.25)^2(0.20)^2 + (-0.25)^2(0.30)^2 + 2(1.25)(-0.25)(0.20)(0.30)(0.20) \quad \Rightarrow \quad \sigma_p^2 = (1.5625)(0.04) + (0.0625)(0.09) + (-0.625)(0.012) = 0.060625$

$\sigma_p = \sqrt{0.060625} \approx 0.2462$ or 24.62%

---

**c. Target volatility $\sigma_p = 0.21$. Let $w = w_{Omega}$ and $(1-w) = w_{Sigma}$.**

$\sigma_p^2 = w^2\sigma_1^2 + (1-w)^2\sigma_2^2 + 2w(1-w)\rho\sigma_1\sigma_2$

$0.21^2 = w^2(0.20)^2 + (1-w)^2(0.30)^2 + 2w(1-w)(0.20)(0.20)(0.30) \quad \Rightarrow \quad 0.0441 = 0.04w^2 + 0.09(1 - 2w + w^2) + 0.024(w - w^2)$

$0.0441 = 0.04w^2 + 0.09 - 0.18w + 0.09w^2 + 0.024w - 0.024w^2 \quad \Rightarrow \quad 0.106w^2 - 0.156w + 0.0459 = 0$

Using the quadratic formula $w = \frac{-b \pm \sqrt{b^2 - 4ac}}{2a}$:

$w = \frac{0.156 \pm \sqrt{(-0.156)^2 - 4(0.106)(0.0459)}}{2(0.106)} = \frac{0.156 \pm \sqrt{0.024336 - 0.0194616}}{0.212} \quad \Rightarrow \quad w = \frac{0.156 \pm 0.069817}{0.212}$

Two possible weights for Omega:
1. $w_{Omega} \approx 1.0652$ (implies $w_{Sigma} \approx -0.0652) \quad \Rightarrow \quad E(R_p) = (1.0652 \times 0.10) + (-0.0652 \times 0.15) \approx 9.67\%$

2. $w_{Omega} \approx 0.4065$ (implies $w_{Sigma} \approx 0.5935) \quad \Rightarrow \quad E(R_p) = (0.4065 \times 0.10) + (0.5935 \times 0.15) \approx 12.97\%$

An rational investor would choose Set 2. While both portfolios have the same 21% risk, Set 2 provides a higher expected return, placing it on the Efficient Frontier. Set 1 is inefficient as it provides less return for the same level of risk.

### Question 2.4

You expect Kappa Corporation stock to have a 20% return next year with a 45% volatility. 
*   You have $100,000 of your own capital to invest, but you plan to leverage your position by investing a total of $140,000 in Kappa. You will raise the additional $40,000 by short-selling either Lambda Corporation or Mu Corporation stock. 

*   Both Lambda and Mu have an expected return of 10% and a volatility of 30%. 
*   Lambda has a correlation of +0.6 with Kappa.
*   Mu has a correlation of -0.6 with Kappa.

Which stock should you choose to short to minimize your portfolio's total risk?

---

### Question 2.4: Solution

The objective is to minimize portfolio volatility. Since the expected return of the shorted stock is the same in both scenarios ($10\%$), the expected return of the total portfolio will be identical; only the risk profile changes based on the correlation.

**Portfolio Weights (Total Value = $100,000):**
*   $w_{Kappa} = \$140,000 / \$100,000 = 1.40 \quad \Rightarrow \quad w_{Short} = -\$40,000 / \$100,000 = -0.40$

**Scenario 1: Short Lambda (Correlation = +0.6)**

$\sigma_p^2 = (1.40)^2(0.45)^2 + (-0.40)^2(0.30)^2 + 2(1.40)(-0.40)(0.45)(0.30)(0.60) \quad \Rightarrow \quad \sigma_p^2 = (1.96)(0.2025) + (0.16)(0.09) - 0.09072 \quad \Rightarrow \quad \sigma_p^2 = 0.32058$

$\sigma_p = \sqrt{0.32058} \approx 0.5662 $

**Scenario 2: Short Mu (Correlation = -0.6)**

$\sigma_p^2 = (1.40)^2(0.45)^2 + (-0.40)^2(0.30)^2 + 2(1.40)(-0.40)(0.45)(0.30)(-0.60) \quad \Rightarrow \quad \sigma_p^2 = (1.96)(0.2025) + (0.16)(0.09) + 0.09072 \quad \Rightarrow \quad \sigma_p^2 = 0.50202$

$\sigma_p = \sqrt{0.50202} \approx 0.7085$


The portfolio volatility is significantly lower when shorting Lambda Corporation (0.5662 vs 0.7085). 

When you are long one asset and short another, a positive correlation provides a diversification benefit.

### Question 2.5

You are the manager of a regional investment fund. The fund currently holds a portfolio consisting of two stocks, Rho and Sigma, plus $100 million in risk-free Treasury bills. The fund’s holdings in Rho and Sigma are currently valued at $120 million and $180 million, respectively.

Your research team has provided the following data:

| Asset | Annual Standard Deviation | Beta |
| :--- | :---: | :---: |
| Stock Rho | 30% | 1.2 |
| Stock Sigma | 40% | 0.8 |
| Market Portfolio | 20% |  |

The correlation coefficient between returns on Rho and Sigma is 0.5. The risk-free interest rate is 5%, and the expected return on the market portfolio is 11%.

a. Calculate the standard deviation of the returns for the portfolio currently held by the fund.

b. Calculate the beta of the fund's portfolio.

c. Assuming the CAPM holds, calculate the expected return of the fund's portfolio.

d. Is the current portfolio efficient under the CAPM? If not, identify the weightings of a complete portfolio that would provide the same expected return but with a lower level of risk. What is the standard deviation of this portfolio?

---

### Question 2.5: Solution

Total portfolio value = $120M (Rho) + $180M (Sigma) + $100M (Risk-Free) = $400M.

$w_{Rho} = 120 / 400 = 0.30$

$w_{Sigma} = 180 / 400 = 0.45$

$w_{rf} = 100 / 400 = 0.25$

**a. Portfolio Standard Deviation:**
The risk-free asset has zero variance and zero correlation with the stocks. The portfolio variance is:

$\sigma_p^2 = w_{Rho}^2\sigma_{Rho}^2 + w_{Sigma}^2\sigma_{Sigma}^2 + 2w_{Rho}w_{Sigma}\sigma_{Rho}\sigma_{Sigma}\rho_{Rho,Sigma}$

$\sigma_p^2 = (0.30)^2(0.30)^2 + (0.45)^2(0.40)^2 + 2(0.30)(0.45)(0.30)(0.40)(0.50) \quad \Rightarrow \quad \sigma_p^2 = 0.0567$

$\sigma_p = \sqrt{0.0567} \approx 0.2381$ 

**b. Portfolio Beta:**
The portfolio beta is the weighted average of the asset betas:

$\beta_p = w_{Rho}\beta_{Rho} + w_{Sigma}\beta_{Sigma} + w_{rf}\beta_{rf}$

$\beta_p = (0.30)(1.2) + (0.45)(0.8) + (0.25)(0) = 0.36 + 0.36 + 0 = 0.72$

**c. Expected Return (CAPM):**

$E(R_p) = R_f + \beta_p[E(R_M) - R_f]$

$E(R_p) = 0.05 + 0.72(0.11 - 0.05) = 0.05 + 0.72(0.06) = 0.05 + 0.0432 = 0.0932$

**d. Efficiency Analysis:**
To find an efficient portfolio with the same expected return (9.32%), we combine the market portfolio and the risk-free asset:

$E(R_{eff}) = w_M E(R_M) + (1 - w_M)R_f$

$0.0932 = w_M(0.11) + (1 - w_M)(0.05) \quad \Rightarrow \quad w_M = 0.72$

The efficient portfolio has a 72% weight in the market and a 28% weight in the risk-free asset.

$\sigma_{eff} = w_M \sigma_M = 0.72 \times 0.20 = 0.1440$

*The fund's current portfolio is not efficient. An efficient portfolio with the same expected return (9.32%) has a standard deviation of only 14.40%, which is significantly lower than the fund's current risk of 23.81%.*

### Question 2.6

Epsilon Processing operates a specialized manufacturing plant.
- On average, annual revenues from the facility are $120 million.

- Currently, all of the facility’s operating costs are variable and consistently represent 70% of revenues. Power costs constitute exactly one-fourth of these total variable costs, averaging $21 million per year.

- The facility has an asset beta ($\beta_U$) of 1.40. The current risk-free rate is 4%, and the market risk premium is 5%. The corporate tax rate is 25%. There are no other costs or growth expected.

a. Estimate the total value of the facility today.

b. Suppose the company enters a long-term contract to supply all of the facility’s power needs for a fixed cost of $15 million per year (pre-tax). What is the value of the facility if this contract is signed?

c. How would signing the contract in (b) change the facility’s implied unlevered cost of capital? Comment on why the overall risk profile of the net cash flows has changed.

---


### Question 2.6: Solution

Calculate the unlevered free cash flow (FCF) and the required return ($r_U$) using the CAPM.

*   $\text{EBIT} = \text{Revenue} - \text{Total Variable Costs} = \$120\text{M} - (0.70 \times \$120\text{M}) = \$36,000,000$
*   $\text{FCF} = \text{EBIT} \times (1 - T_c) = \$36,000,000 \times 0.75 = \$27,000,000$
*   $r_U = R_f + \beta_U(\text{MRP}) = 0.04 + 1.40(0.05) = 0.11$ (or 11.0%)
*   $V_U = \frac{\text{FCF}}{r_U} = \frac{\$27,000,000}{0.11} \approx \$245,454,545$

The contract replaces a risky variable cost with a certain fixed cost. Because the cash flows are no longer homogeneous in risk, we value them separately.

**PV of Old Variable Power Costs:** These were risky and correlated with revenues, so they are discounted at the baseline $r_U$ (11%).
$$\text{After-tax Variable Cost} = \$21,000,000 \times (1 - 0.25) = \$15,750,000$$
$$\text{PV(Old Costs)} = \frac{\$15,750,000}{0.11} \approx \$143,181,818$$

**PV of New Fixed Power Costs:** These are now certain (contractually fixed), so they are discounted at the risk-free rate $R_f$ (4%).
$$\text{After-tax Fixed Cost} = \$15,000,000 \times (1 - 0.25) = \$11,250,000$$
$$\text{PV(New Costs)} = \frac{\$11,250,000}{0.04} = \$281,250,000$$

**Net Impact and New Value:**
$$\text{NPV(Contract)} = \text{PV(Old)} - \text{PV(New)} = \$143,181,818 - \$281,250,000 = -\$138,068,182$$
$$V_{new} = V_U + \text{NPV(Contract)} = \$245,454,545 - \$138,068,182 \approx \$107,386,363$$

**Implied New Cost of Capital**

*   New variable costs (excluding power): $\$84\text{M} - \$21\text{M} = \$63,000,000$
*   New fixed costs: $\$15,000,000$
*   $\text{New EBIT} = \$120,000,000 - \$63,000,000 - \$15,000,000 = \$42,000,000$
*   $\text{New FCF} = \$42,000,000 \times 0.75 = \$31,500,000$
*   $r_{U,new} = \frac{\text{New FCF}}{V_{new}} = \frac{\$31,500,000}{\$107,386,363} \approx 29.33\%$

Signing the contract increases the facility's implied unlevered cost of capital to 29.33%. 
- This occurs because the firm has significantly increased its operating leverage. 
- However, this 29.33% represents an aggregate expected return; after the contract, the cash flows should ideally be valued by separating the risky operating components from the fixed contractual obligations.

### Question 2.7

Given the following information for Sigma Corp, find the weighted average cost of capital (WACC). Assume the company’s tax rate is 25 percent.

Debt: 20,000 5.0 percent coupon bonds outstanding, $1,000 par value, 12 years to maturity, selling for 97 percent of par; the bonds make semiannual payments.

Common stock: 600,000 shares outstanding, selling for $40 per share; the beta is 1.25.

Market: 7.0 percent market risk premium and 3.0 percent risk-free rate.

---

### Question 2.7: Solution

Find the Market Value of Equity ($E$) and Debt ($D$):

$$ E = \text{Shares Outstanding} \times \text{Price per Share} = 600,000 \times \$40 = \$24,000,000 $$

The market price of one bond is 97% of its par value, or $\$1,000 \times 0.97 = \$970$.

$$ D = \text{Number of Bonds} \times \text{Bond Price} = 20,000 \times \$970 = \$19,400,000 $$

$$ V = E + D = \$24,000,000 + \$19,400,000 = \$43,400,000 $$

Find the Weights:

$$w_E = \frac{E}{V} = \frac{\$24,000,000}{\$43,400,000} \approx 0.5530 \quad \text{and} \quad w_D = \frac{D}{V} = \frac{\$19,400,000}{\$43,400,000} \approx 0.4470 $$

Find the Cost of Equity ($r_E$) and Cost of Debt ($r_D$):

$$ r_E = R_f + \beta \times (\text{Market Risk Premium}) = 0.03 + 1.25(0.07) = 0.1175 $$

For the cost of debt, solve for the yield to maturity (YTM):

$N = 12 \text{ years} \times 2 = 24 \text{ periods}$; $PV = -\$970$; $PMT = \frac{\$1,000 \times 0.05}{2} = \$25$; $FV = \$1,000$.

$$ 970 = 25 \times \left[ \frac{1 - (1+y)^{-24}}{y} \right] + \frac{1000}{(1+y)^{24}} \quad \Rightarrow \quad y \approx 2.6732\% $$
$$ r_D = 2.6732\% \times 2 = 0.053464 \approx 5.35\% $$

$$ WACC = w_E r_E + w_D r_D (1 - T_c) $$

$$ WACC = (0.5530)(0.1175) + (0.4470)(0.053464)(1 - 0.25) \quad \Rightarrow \quad WACC = 0.0829033 \approx 8.29\% $$

### Question 2.8

The Omega Fund (Omega) has an expected return of 15.0% and a volatility of 18.0%. Omega claims that no other portfolio offers a higher Sharpe ratio. Suppose this claim is true, and the risk-free interest rate is 3.0%.

a. What is Omega's Sharpe Ratio?

b. If Delta stock has a volatility of 35.0% and an expected return of 7.0%, what must be its correlation with the Omega Fund?

c. If the Sigma Fund (Sigma) has a correlation of 75% with the Omega Fund, what is the Sharpe ratio of the Sigma Fund?

---

### Question 2.8: Solution

**a. Omega's Sharpe Ratio**
$$ SR_{Omega} = \frac{E[R_{Omega}] - r_f}{\sigma(R_{Omega})} = \frac{0.1500 - 0.0300}{0.1800} = \frac{0.1200}{0.1800} \approx 0.6667 $$

**b. Implied Correlation for Delta Stock**
If Omega is the efficient portfolio, then any other asset must satisfy the CAPM relationship relative to Omega:
$$ E[R_{Delta}] = r_f + \beta_{Delta}^{Omega} (E[R_{Omega}] - r_f) $$
$$ 0.0700 = 0.0300 + \beta_{Delta}^{Omega} (0.1500 - 0.0300) \implies \beta_{Delta}^{Omega} = \frac{0.0400}{0.1200} \approx 0.3333 $$

Use the definition of beta to find the correlation:
$$ \beta_{Delta}^{Omega} = \frac{Corr(R_{Delta}, R_{Omega}) \times \sigma_{Delta}}{\sigma_{Omega}} $$
$$ 0.3333 = \frac{Corr(R_{Delta}, R_{Omega}) \times 0.3500}{0.1800} \implies Corr(R_{Delta}, R_{Omega}) = \frac{0.3333 \times 0.1800}{0.3500} \approx 0.1714 $$

**c. Sharpe Ratio of the Sigma Fund**

Assuming the Sigma Fund is priced according to CAPM relative to the efficient Omega Fund, its expected return is:
$$ E[R_{Sigma}] = r_f + \beta_{Sigma}^{Omega} (E[R_{Omega}] - r_f) $$

$$ \beta_{Sigma}^{Omega} = \frac{Corr(R_{Sigma}, R_{Omega}) \times \sigma_{Sigma}}{\sigma_{Omega}} = \frac{0.7500 \times \sigma_{Sigma}}{0.1800} $$

The excess return is:
$$ E[R_{Sigma}] - 0.0300 = \left( \frac{0.7500 \times \sigma_{Sigma}}{0.1800} \right) \times (0.1200) $$
$$ E[R_{Sigma}] - 0.0300 = \sigma_{Sigma} \times \left( \frac{0.7500 \times 0.1200}{0.1800} \right) = \sigma_{Sigma} \times 0.5000 $$

The Sharpe Ratio is the excess return divided by volatility:
$$ SR_{Sigma} = \frac{E[R_{Sigma}] - r_f}{\sigma_{Sigma}} = \frac{\sigma_{Sigma} \times 0.5000}{\sigma_{Sigma}} = 0.5000 $$

### Question 3.1

Omega Commodities enters into a short gold futures contract when the futures price is $2,000 per ounce. The contract size is 100 ounces. To open the position, the firm must deposit an initial margin of $12,000, and the exchange requires a maintenance margin of $9,000.

1. Calculate the gain or loss for Omega Commodities if the gold price at the end of the contract is:
    * (a) $1,980 per ounce
    * (b) $2,030 per ounce

2. Determine the specific price level that would trigger a margin call for this short position.

3. At what gold price could Omega Commodities withdraw exactly $2,500 from its margin account?

---

### Question 3.1: Solution

**1. Profit/Loss Calculation**

$$ \text{Profit/Loss} = \text{Contract Size} \times (F_0 - S_T) $$

*   **Case (a): Final price is $1,980**
    $$ \text{Gain} = 100 \times (\$2,000 - \$1,980) = 100 \times \$20 = \$2,000 $$

*   **Case (b): Final price is $2,030**
    $$ \text{Loss} = 100 \times (\$2,000 - \$2,030) = 100 \times (-\$30) = -\$3,000 $$

**2. Margin Call Price**

A margin call occurs when the account balance falls below the maintenance margin. For a short position, this happens as the price increases.
$$ \text{Account Balance} = \text{Initial Margin} + \text{Quantity} \times (F_0 - P) $$
$$ 12,000 + 100 \times (2,000 - P) < 9,000 $$
$$ 100 \times (2,000 - P) < -3,000 \implies 2,000 - P < -30 \implies P > \$2,030 $$

*A margin call will be triggered if the gold price rises above $2,030.*

**3. Withdrawal Price**

Funds can be withdrawn when the account balance exceeds the initial margin. To withdraw $2,500, the total balance must reach $14,500 ($12,000 + $2,500).
$$ 12,000 + 100 \times (2,000 - P) = 14,500 $$
$$ 100 \times (2,000 - P) = 2,500 \implies P = \$1,975 $$

*Omega Commodities can withdraw $2,500 if the gold price drops to $1,975.*

### Question 3.4

The current price of Sigma Corp stock is $60. 
-   In the next year, the stock price will either increase by 20% or decrease by 20%. 

-   The stock pays no dividends. The one-year risk-free interest rate is 4%. 

-   Using the Binomial Model, calculate the price of a one-year European put option on Sigma Corp stock with a strike price of $64.

---

### Question 3.4: Solution

- Current stock price $S_0 = \$60$.
- Strike price $K = \$64$.
- Risk-free rate $r_f = 0.04$, so the risk-free factor is $R = 1.04$.
- Up-factor $u = 1.20$ and down-factor $d = 0.80$.

$S_u = S_0 \times u = \$60 \times 1.20 = \$72.00$.
$f_u = \max(K - S_u, 0) = \max(\$64 - \$72, 0) = \$0$.

$S_d = S_0 \times d = \$60 \times 0.80 = \$48.00$.
$f_d = \max(K - S_d, 0) = \max(\$64 - \$48, 0) = \$16$.

**Method 1: Replicating Portfolio**

$$ \Delta = \frac{f_u - f_d}{S_u - S_d} = \frac{\$0 - \$16}{\$72 - \$48} = \frac{-\$16}{\$24} \approx -0.6667 $$

$$ \Delta S_u + B \times R = f_u \implies (-0.666667)(72) + 1.04B = 0 $$

$$ -48 + 1.04B = 0 \implies B = \frac{48}{1.04} \approx \$46.1538 $$

The price of the put today is the cost of this replicating portfolio.

$$ P_0 = \Delta S_0 + B = (-0.666667)(\$60) + \$46.1538 = -\$40 + \$46.1538 = \$6.1538 $$

**Method 2: Risk-Neutral Valuation**

$$ p_u = \frac{R - d}{u - d} = \frac{1.04 - 0.80}{1.20 - 0.80} = \frac{0.24}{0.40} = 0.6000 $$

$$p_d = 1 - 0.6000 = 0.4000$$

The price of the put today is the discounted expected payoff using risk-neutral probabilities:

$$ P_0 = \frac{p_u \times f_u + p_d \times f_d}{R}  \quad \Rightarrow \quad P_0 = \frac{(0.6000)(\$0) + (0.4000)(\$16)}{1.04} = \frac{\$6.40}{1.04} \approx \$6.1538 $$

### Question 3.5
Omega Corp’s current stock price is $100. Suppose that over the next year, the stock price will either increase by 40% or decrease by 20%. The annual risk-free rate is 5%.
