### Question 1.1

Consider an offer to trade 10 troy ounces (oz) of platinum for 800 oz of silver. Current competitive market prices are approximately £1,100 per oz for platinum and £15 per oz for silver. 

* Based on current market prices, does this trade represent a positive net value? Explain using the Law of One Price or valuation principles.

* Under what specific market price conditions (deviation from current prices) might this trade become unappealing?

* What action might you consider if you strongly anticipated silver prices falling to £13 per oz within the next day, assuming you could execute related transactions instantly?

* How does your personal belief about the "fairness" of the current platinum or silver prices affect the decision to accept the trade based solely on market values?

---

### Question 1.1: Solution

To determine if the trade has a positive net value, we compare the market value of what is received to the market value of what is given up:
*   **Value Received (Silver):** $800 \text{ oz} \times £15/\text{oz} = £12,000$
*   **Value Given (Platinum):** $10 \text{ oz} \times £1,100/\text{oz} = £11,000$
*   **Net Value:** $£12,000 - £11,000 = +£1,000$

Since the Net Value is positive, the trade is financially advantageous. According to the **Law of One Price**, the value of the trade is determined by current market prices in competitive markets, regardless of the specific assets involved.

The trade becomes unappealing (Net Value $\leq 0$) if the value of the platinum meets or exceeds the value of the silver. This occurs if:
*   The price of silver drops to £13.75 or lower ($£11,000 / 800 \text{ oz}$).
*   The price of platinum rises to £1,200 or higher ($£12,000 / 10 \text{ oz}$).

If you expect silver to drop to £13 tomorrow:
*   The future value of the silver you receive would be $800 \times £13 = £10,400$. 
*   This is less than the £11,000 value of the platinum you are giving up.
*   If you can execute transactions instantly, you should accept the trade today to get the silver at its current £15 valuation and sell the silver immediately for £12,000 in cash. This locks in the £1,000 profit today and removes the risk of the price drop.

**Personal Belief vs. Market Value**

In competitive financial markets, personal beliefs about "fairness" are irrelevant to the valuation. 
* The market price represents the consensus value at which you can actually buy or sell the assets. 
* If the market value of the silver is higher than the platinum, the trade creates wealth today, even if you personally believe silver is "overpriced."

### Question 1.2

Consider a risk-free investment note issued by the Sigma Venture Group that promises to pay investors the following guaranteed cash flows:

 * $\$100$ exactly one year from today
   
 * $\$150$ exactly two years from today

 * $\$200$ exactly three years from today

Suppose the current market interest rates for these specific risk-free maturities are as follows:
    
 * One-year risk-free rate ($r_1$): 5.0%
 * Two-year risk-free rate ($r_2$): 5.5%
 * Three-year risk-free rate ($r_3$): 6.0%

* What is the theoretical "no-arbitrage" price of this note today? 

* If this security is currently trading in the market for a price of $\$410$, explain why an arbitrage opportunity exists. 
    * Describe the specific sequence of transactions you would undertake today to lock in a risk-free profit without using any of your own money. 
    * Calculate the total profit realized today.

---

### Question 1.2: Solution

**Theoretical No-Arbitrage Price**

$$PV = \frac{C_1}{(1+r_1)^1} + \frac{C_2}{(1+r_2)^2} + \frac{C_3}{(1+r_3)^3}$$

$$PV = \frac{\$100}{1.05^1} + \frac{\$150}{1.055^2} + \frac{\$200}{1.06^3}$$

$$PV = \$95.238 + \$134.768 + \$167.924 = \$397.93$$

**Arbitrage Strategy and Profit**

*   The market price ($\$410.00$) is higher than the theoretical price ($\$397.93$). The security is **overpriced**. 

*   Sell the overpriced security and "replicate" the future obligations by investing in the bank (lending) at the lower theoretical cost.

| Action | Today ($t=0$) | Year 1 | Year 2 | Year 3 |
| :--- | :---: | :---: | :---: | :---: |
| **Sell the Security** | $+\$410.00$ | $-\$100.00$ | $-\$150.00$ | $-\$200.00$ |
| **Invest in Bank (Lend)** | $-\$397.93$ | $+\$100.00$ | $+\$150.00$ | $+\$200.00$ |
| **Net Cash Flow** | **$+\$12.07$** | **$\$0.00$** | **$\$0.00$** | **$\$0.00$** |

*By selling the security and investing the required present value in the bank, you realize a risk-free profit of **$\$12.07$ today** with zero net cash flow obligations in the future.*

### Question 2.2

*   Eternal Capital Advisors is offering an investment policy that promises to pay you and your heirs $\$25,000$ per year forever.
    *   If the required return on this type of investment is 5.8%, what is the maximum amount you should be willing to pay for this policy today?
    
    *   Suppose the company lists the price of this policy at $\$450,000$. At what interest rate would this be considered a "fair" price?
    
    *   Alternatively, assume the firm offers a different version of the policy that lasts for only 30 years. However, the first payment is $\$25,000$ (received in one year), and the payments grow by 3% every year thereafter. If the required return remains 5.8%, what is the present value of this growing annuity?

---

### Question 2.2: Solution

**Present Value (PV) of the Perpetuity**

$$PV = \frac{C}{r} \quad \Rightarrow \quad PV = \frac{\$25,000}{0.058} = \$431,034.48$$

**The Interest Rate for a $\$450,000$ Price**

$$r = \frac{C}{PV} \quad \Rightarrow \quad r = \frac{\$25,000}{\$450,000} \approx 0.05556 \text{ or } 5.56\%$$

**Present Value of the Growing Annuity**

$$PV = \frac{C_1}{r - g} \left[ 1 - \left( \frac{1 + g}{1 + r} \right)^n \right]$$

$$PV = \frac{\$25,000}{0.058 - 0.03} \times \left[ 1 - \left( \frac{1 + 0.03}{1 + 0.058} \right)^{30} \right] \quad \Rightarrow \quad PV = \frac{\$25,000}{0.028} \times \left[ 1 - (0.973535)^{30} \right] \\ \quad \Rightarrow \quad PV = \$892,857.14 \times 0.553132 \approx \$493,868.04$$

### Question 2.3
Investment X offers to pay you $4,000 per year for 20 years, whereas Investment Y offers to pay you $9,000 per year for 6 years. Which investment would you choose under the following interest rate scenarios?

*   4 percent
*   18 percent

---

### Question 2.3: Solution

$$PV = C \times \left[ \frac{1 - (1 + r)^{-n}}{r} \right]$$

**Interest Rate = 4%**

Investment X (n=20, C=4,000):
$$PV_X = 4,000 \times \left[ \frac{1 - (1.04)^{-20}}{0.04} \right] \approx 4,000 \times 13.5903 \approx \$54,361.20$$

Investment Y (n=6, C=9,000):
$$PV_Y = 9,000 \times \left[ \frac{1 - (1.04)^{-6}}{0.04} \right] \approx 9,000 \times 5.2421 \approx \$47,178.90$$

**Interest Rate = 18%**

Investment X (n=20, C=4,000):
$$PV_X = 4,000 \times \left[ \frac{1 - (1.18)^{-20}}{0.18} \right] \approx 4,000 \times 5.3527 \approx \$21,410.80$$

Investment Y (n=6, C=9,000):
$$PV_Y = 9,000 \times \left[ \frac{1 - (1.18)^{-6}}{0.18} \right] \approx 9,000 \times 3.4976 \approx \$31,478.40$$

### Question 2.4
* You have just finalized the purchase of a new luxury condominium. To finance the transaction, you have secured a 25-year mortgage for 75 percent of the $750,000 purchase price. The monthly payment on this mortgage is fixed at $3,500.
  - What is the Annual Percentage Rate (APR) on this loan?
  
  - What is the Effective Annual Rate (EAR) on this loan?

---

### Question 2.4: Solution

Using the Present Value of an Annuity formula:
$$PV = PMT \times \left[ \frac{1 - (1 + i)^{-n}}{i} \right] \quad \Rightarrow \quad 562,500 = 3,500 \times \left[ \frac{1 - (1 + i)^{-300}}{i} \right] \quad \Rightarrow \quad i \approx 0.004674 \text{ (or 0.4674\% per month)}$$

The APR is the periodic rate multiplied by the number of periods in a year:
$$APR = i \times 12 \quad \Rightarrow \quad APR = 0.4674\% \times 12 = 5.6088\%$$

The EAR accounts for the effect of compounding over the year:
$$EAR = (1 + i)^{12} - 1 \quad \Rightarrow \quad EAR = (1.004674)^{12} - 1 \quad \Rightarrow \quad EAR = 1.057551 - 1 = 0.057551 \text{ or } 5.76\%$$

### Question 2.5

* Suppose a company takes out a $400,000 commercial mortgage with a 9% APR to be repaid in monthly installments over a 20-year term.
    
    * Calculate the required monthly payment for this mortgage.
    
    * If the company decides to pay off the entire remaining balance of the loan immediately after making the 96th monthly payment (at the end of Year 8), what will the payoff amount be?

    * Assume the firm wants to fully amortize the loan by the end of the 20th year but plans to make two one-time balloon payments of $60,000 at the end of Year 5 (month 60) and Year 12 (month 144). What would the new fixed monthly payment be under this arrangement?

    * The firm's CFO is exploring a shorter 12-year (144-month) mortgage. The firm can only afford a maximum monthly payment of $3,800. To ensure the loan is fully paid off by the end of Year 12, the firm will make two equal balloon payments at the end of Year 4 (month 48) and Year 9 (month 108).
        * What is the minimum amount required for each of these two equal balloon payments to fully amortize the loan?

---

### Question 2.5: Solution

$$PMT = 400,000 \times \frac{0.0075}{1 - (1.0075)^{-240}} \approx 3,598.90$$

**Remaining balance after 96 payments**

This is the present value of the 144 remaining payments ($240 - 96 = 144$):
$$PV_{96} = 3,598.9038 \times \left[ \frac{1 - (1.0075)^{-144}}{0.0075} \right] \approx 316,239.61$$

**Monthly payments with $60,000 balloons at months 60 and 144**

Calculate the present value (at $t=0$) of the balloon payments:
$$PV_{B1} = \frac{60,000}{(1.0075)^{60}} \approx 38,321.98 \quad \text{and} \quad PV_{B2} = \frac{60,000}{(1.0075)^{144}} \approx 20,458.01$$

Subtract these from the loan amount and solve for the new $PMT$ over 240 months:
$$400,000 - 38,321.98 - 20,458.01 = PMT \times \left[ \frac{1 - (1.0075)^{-240}}{0.0075} \right] \quad \Rightarrow \quad 341,220.01 = PMT \times 111.144954 \quad \Rightarrow \quad PMT \approx 3,070.04$$

**Minimum equal balloon payments for a 144-month loan capped at $3,800/month**

First, find the PV of the $3,800 annuity for 144 months:
$$PV_{annuity} = 3,800 \times \left[ \frac{1 - (1.0075)^{-144}}{0.0075} \right] \approx 333,910.15$$

The balloons ($B$) at $t=48$ and $t=108$ must cover the remaining balance:
$$400,000 - 333,910.15 = \frac{B}{(1.0075)^{48}} + \frac{B}{(1.0075)^{108}} \quad \Rightarrow \quad 66,089.85 = B \times (0.698579 + 0.446210) \quad \Rightarrow \quad B \approx 57,729.53$$

### Question 2.6
Maximilian Weber owns "Kraftwerk Fitness," a popular boutique gym in Berlin, Germany. He currently charges members an annual fee of €400 and has a steady base of 700 members. Maximilian plans to increase the annual fee by 4% each year and expects his membership count to grow at a constant rate of 5% per year over the next five years. His business incurs operating expenses of €150,000 per year, which are projected to increase at an inflation rate of 2.5% annually. At the end of five years, Maximilian intends to invest €500,000 into a high-end sports equipment franchise and then close his gym to begin a 20-year retirement.

After the equipment investment, the remaining savings will be held in an account earning a 7.5% annual nominal return. Maximilian plans to fund his retirement by withdrawing an equal amount in real terms at the beginning of each year, depleting the account completely by the end of the 20 years.

What is the annual amount Maximilian can spend during his retirement expressed in today’s euros (Year 0), and what is the corresponding nominal amount of his final withdrawal, assuming the withdrawals are indexed to inflation?

---


### Question 2.6: Solution

**Net Cash Flows and Accumulation (Years 1–5)**
*   Combined Revenue Growth Rate ($g_{rev}$): $(1.04 \times 1.05) - 1 = 0.092$ or $9.2\%$
*   Expense Inflation Rate ($i$): 2.5%
*   Investment Return ($r$): 7.5%
*   Initial Revenue ($R_0$): $700 \times €400 = €280,000$

| Year ($t$) | Revenue: €280k $\times (1.092)^t$ | Expenses: €150k $\times (1.025)^t$ | Net Cash Flow |
| :--- | :--- | :--- | :--- |
| 1 | €305,760.00 | €153,750.00 | €152,010.00 |
| 2 | €333,889.92 | €157,593.75 | €176,296.17 |
| 3 | €364,607.79 | €161,533.59 | €203,074.20 |
| 4 | €398,151.71 | €165,571.93 | €232,579.78 |
| 5 | €434,781.67 | €169,711.23 | €265,070.44 |

Calculate the Future Value (FV) of these savings at $t=5$ using the 7.5% return:
$$FV = 152,010(1.075)^4 + 176,296.17(1.075)^3 + 203,074.20(1.075)^2 + 232,579.78(1.075)^1 + 265,070.44$$
$$FV = 203,004.66 + 219,012.14 + 234,677.56 + 250,023.26 + 265,070.44 = €1,171,788.06$$

**Retirement Starting Balance:** At $t=5$, Maximilian makes his franchise investment,
$$Balance_{t=5} = €1,171,788.06 - €500,000 = €671,788.06$$

**Annual Retirement Spending (Real terms):** To determine constant purchasing power, we use the real interest rate ($r_{real}$),
$$r_{real} = \frac{1 + 0.075}{1 + 0.025} - 1 = 0.04878049$$

Since withdrawals are made at the **beginning of each year** (Annuity Due) for 20 years:
$$PV = W_{t=5} \times \left[ \frac{1 - (1 + r_{real})^{-20}}{r_{real}} \right] \times (1 + r_{real})$$
$$671,788.06 = W_{t=5} \times 12.59212 \times 1.04878 \quad \Rightarrow \quad W_{t=5} \approx €50,868.63$$

In today's euros (Year 0):
$$W_{today} = \frac{€50,868.63}{(1.025)^5} \approx \mathbf{€44,960.46}$$

**Final Nominal Withdrawal:** Maximilian retires for 20 years. The first withdrawal is at $t=5$. The 20th (final) withdrawal occurs at $t=24$.
$$W_{final} = W_{t=5} \times (1 + i)^{19} \quad \Rightarrow \quad W_{final} = €50,868.63 \times (1.025)^{19} \approx \mathbf{€81,321.14}$$

### Question 3.1

*   AeroTech Dynamics is evaluating a proposal to build a new regional assembly line. This expansion is projected to generate additional net cash inflows of $\$450,000$ for the firm during the first year after completion. These cash flows are expected to grow steadily at a rate of 6 percent per year indefinitely. The project requires an upfront investment of $\$5,200,000$.
    
    *   If AeroTech Dynamics requires a return of 12 percent on such capital projects, should the assembly line expansion be undertaken? 
    
    *   The company's management is concerned about the long-term growth forecast. At what constant perpetual growth rate would this expansion project just break even, assuming the company still requires a 12 percent return on investment?

---

### Question 3.1: Solution


$$PV = \frac{C_1}{r - g} \quad \Rightarrow \quad PV = \frac{\$450,000}{0.12 - 0.06} = \frac{\$450,000}{0.06} = \$7,500,000$$

$$NPV = -Initial \ Investment + PV \quad \Rightarrow \quad NPV = -\$5,200,000 + \$7,500,000 = \$2,300,000$$

**Break-even Growth Rate**

To find the growth rate where the project breaks even, we set the NPV to zero:
$$NPV = -Investment + \frac{C_1}{r - g} = 0 \quad \Rightarrow \quad \$5,200,000 = \frac{\$450,000}{0.12 - g}$$

$$\Rightarrow \quad (0.12 - g) \times \$5,200,000 = \$450,000 \quad \Rightarrow \quad g = 0.033462$$

### Question 3.2

*   A commercial property investment group is deciding between two mutually exclusive redevelopment opportunities in a growing tech hub. Both projects require an identical upfront investment of $20 million.
    *   **Project Constant (PC)**: This project is a stabilized office complex expected to generate a flat net cash flow of $3.2 million per year in perpetuity, starting one year from today.
    
    *   **Project Growth (PG)**: This project is a modern retail and residential hub expected to generate $2.4 million at the end of the first year. This income stream is projected to grow at a constant rate of 3.5% per year indefinitely.

    
    *   Calculate the Internal Rate of Return (IRR) for both projects. Which project has the higher IRR?
    
    *   Calculate the Net Present Value (NPV) for both projects assuming the firm's cost of capital is 9%. Which project should be chosen based on the NPV rule?
    
    *   Calculate the crossover rate (the discount rate at which both projects have the same NPV). 
    
    *   Based on your analysis, for what range of cost of capital would the IRR rule lead to the correct decision for these mutually exclusive projects? Provide a detailed range analysis.

---

### Question 3.2: Solution

**IRR for each project**

For a perpetuity: $PV = \frac{C}{r} \Rightarrow r = \frac{C}{PV}$
*   **IRR (Project Constant):**
    $$r_{PC} = \frac{3.2}{20} = 0.16 \text{ or } 16.0\%$$

For a growing perpetuity: $PV = \frac{C_1}{r - g} \Rightarrow r = \frac{C_1}{PV} + g$
*   **IRR (Project Growth):**
    $$r_{PG} = \frac{2.4}{20} + 0.035 = 0.12 + 0.035 = 0.155 \text{ or } 15.5\%$$

Project Constant (PC) has the higher IRR (16.0%).

**NPV at a 9% cost of capital**

*   **NPV (Project Constant):**
    $$NPV_{PC} = -20 + \frac{3.2}{0.09} \approx -20 + 35.56 = \$15.56 \text{ million}$$

*   **NPV (Project Growth):**
    $$NPV_{PG} = -20 + \frac{2.4}{0.09 - 0.035} = -20 + \frac{2.4}{0.055} \approx -20 + 43.64 = \$23.64 \text{ million}$$

**The Crossover Rate**

The crossover rate is the discount rate ($r$) where $NPV_{PC} = NPV_{PG}$:
$$\frac{3.2}{r} = \frac{2.4}{r - 0.035} \quad \Rightarrow \quad r = \frac{0.112}{0.8} = 0.14 \text{ or } 14.0\%$$

**Range Analysis for IRR Rule Reliability**

The IRR rule correctly identifies the best project only when the project with the higher IRR also provides the highest NPV.

| Cost of Capital Range | Project with Higher NPV | Project with Higher IRR | IRR Rule Decision | Reliability of IRR Rule |
| :--- | :---: | :---: | :---: | :---: |
| $r < 14\%$ | PG | PC | Select PC | **Incorrect** (Conflict) |
| $14\% < r < 16\%$ | PC | PC | Select PC | **Correct** |
| $r > 16\%$ | None (Both negative) | PC | Select PC* | **Incorrect** (NPV < 0) |

*For $r > 16\%$, the IRR rule technically suggests PC is "better" than PG, but the NPV rule dictates that neither project should be undertaken.*

*The IRR rule leads to the correct selection between these two mutually exclusive projects only when the cost of capital is between **14% and 16%**.

### Question 3.3

Apex Energy Solutions is evaluating a new high-output hydroelectric dam project. The project involves two distinct financial decisions that must be analyzed using NPV and IRR rules.

**Decision 1: The Hydroelectric Plant Investment**
The construction of the dam requires an immediate investment of $220 million (Year 0). Once operational, the plant is expected to generate net cash inflows of $32 million per year for the next 12 years (Years 1–12). However, starting in Year 13, the firm is legally required to pay for ongoing river ecosystem monitoring and silt management, estimated to cost $2.5 million per year in perpetuity. Assume the firm's cost of capital is 8%.

**Decision 2: The Maintenance Deferral Strategy**
During the first year of operation (Year 1), the firm has the option to skip a non-essential system optimization treatment that costs $10 million. If they forgo this expenditure, they will save $10 million at the end of Year 1, but the resulting wear and tear will require a mandatory specialized repair costing $11.2 million at the end of Year 2.

1.  **Analyze Decision 1:**
    *   Calculate the Net Present Value (NPV) of the hydroelectric plant.
    *   Identify the number of sign changes in the cash flow stream. What does this imply about the number of potential Internal Rates of Return (IRRs)?
    *   Can the IRR rule be used reliably to decide whether to accept or reject the plant?

2.  **Analyze Decision 2:**
    *   Calculate the IRR associated with the specific decision to forgo the Year 1 optimization treatment.
    *   Is this a "lending" or a "borrowing" type of project? 
    *   Based on the 8% cost of capital, should the firm skip the maintenance or pay for it in Year 1? Explain using the IRR rule for this specific case.

---

### Question 3.3: Solution

**Analysis of the Hydroelectric Plant**

*   **PV of Inflows (Years 1–12):**
    $$PV_{inflows} = 32 \times \left[ \frac{1 - (1.08)^{-12}}{0.08} \right] \approx 32 \times 7.536078 = \$241.1545 \text{ million}$$

*   **PV of Perpetual Outflows (Starting Year 13):**
    First, find the value at the end of Year 12: $PV_{12} = \frac{2.5}{0.08} = \$31.25 \text{ million}$.
    
    Discount this value back to today ($t=0$):
    $$PV_{perp,0} = \frac{31.25}{(1.08)^{12}} \approx \frac{31.25}{2.518170} = \$12.4098 \text{ million}$$
    $$NPV = -220 + 241.1545 - 12.4098 \approx \$8.75 \text{million}$$
    

**IRR Reliability:**
*   The cash flow sequence is **(-, +, -)**: An initial outlay, followed by inflows, followed by perpetual outflows.
*   There are two sign changes, which implies there can be up to two IRRs.
*   Because non-conventional cash flows can result in multiple IRRs, the IRR rule is not reliable. 

**Analysis of the Maintenance Deferral**

Forgoing the cost in Year 1 acts as a cash inflow ($+10$), while the required repair in Year 2 is an outflow ($-11.2$).
$$0 = 10 - \frac{11.2}{1 + r} \Rightarrow 10 = \frac{11.2}{1 + r} \Rightarrow 1 + r = 1.12 \Rightarrow \mathbf{IRR = 12\%}$$

**Project Nature and Decision:**
*   This is a borrowing-style project because the firm receives a cash benefit today (by not spending) in exchange for a larger payment in the future.
*   For "borrowing" projects, the IRR rule is inverted: 
    * Accept only if the IRR (borrowing cost) is less than the cost of capital.
    * Since the IRR of 12% is higher than the 8% cost of capital, the firm is effectively borrowing money at a 12% rate, which is more expensive than its 8% cost of capital.
*   The firm should not forgo the maintenance; it should pay the $10 million in Year 1.

### Question 3.5: Solution

### Question 4.1

Precision Logistics Corp. is evaluating the acquisition of an automated inventory sorting system priced at $\$450,000$ with a projected useful life of 6 years. The system will be depreciated on a straight-line basis over its 6-year life toward a book value of zero. The expected market salvage value at the end of Year 6 is $\$50,000$.

By implementing this automation, the firm will be able to eliminate four manual processing roles, resulting in a total pre-tax salary saving of $\$160,000$ per year. 

Furthermore, the increased efficiency of the system will immediately reduce the company's required net working capital (NWC) by $\$75,000$ at the start of the project ($t=0$). This NWC amount will be reinstated (as a cash outflow) when the project concludes and the equipment is sold at the end of Year 6.

The firm's corporate tax rate is 21%, and the required rate of return for this project is 12%.

Should Precision Logistics Corp. proceed with the investment? 

---

### Question 4.1: Solution

$$\text{Annual Depreciation} = \frac{\text{Initial Cost}}{\text{Life}} = \frac{\$450,000}{6} = \$75,000$$

$$OCF = (\text{Savings} - \text{Depreciation}) \times (1 - \tau_c) + \text{Depreciation}$$
$$OCF = (\$160,000 - \$75,000) \times (1 - 0.21) + \$75,000 \quad \Rightarrow \quad OCF = (\$85,000 \times 0.79) + \$75,000 = \$142,150$$

**Incremental Free Cash Flows (FCF)**
*   **Year 0:** Outlay for equipment plus the cash *inflow* from the NWC reduction.
    $$FCF_0 = -\$450,000 + \$75,000 = -\$375,000$$
*   **Years 1–5:** Annual OCF.
    $$FCF_{1-5} = \$142,150$$
*   **Year 6 (Terminal Year):** OCF plus after-tax salvage value (ATSV) minus the NWC reinstatement.
    $$\text{Gain on Sale} = \text{Salvage} - \text{Book Value} = \$50,000 - \$0 = \$50,000$$
    $$ATSV = \$50,000 - (\$50,000 \times 0.21) = \$39,500$$
    $$FCF_6 = \$142,150 + \$39,500 - \$75,000 = \$106,650$$

**NPV at 12%**
$$NPV = -375,000 + \sum_{t=1}^{5} \frac{142,150}{(1.12)^t} + \frac{106,650}{(1.12)^6} \quad \Rightarrow \quad NPV = \$191,451.11$$

Since the NPV is positive ($\$191,451.11$), Precision Logistics Corp. should proceed with the investment.

### Question 4.2

Delta Semiconductor is evaluating the launch of a new specialized microchip fabrication line.

* Based on a Net Present Value (NPV) analysis, should Delta Semiconductor proceed with the fabricated line investment?

$$Annual \ Depreciation = \frac{Initial \ Investment}{Life} = \frac{6,000,000}{5} = 1,200,000$$

$$OCF = (Revenues - Operating \ Costs) \times (1 - \tau_c) + (\tau_c \times Depreciation)$$
$$OCF = (3,500,000 - 1,400,000) \times (1 - 0.25) + (0.25 \times 1,200,000) \quad \Rightarrow \quad OCF = 1,875,000$$

**Free Cash Flows (FCF)**
* Year 0 - Investment in machinery and NWC: $FCF_0 = -6,000,000 - 300,000 = -6,300,000$
* Years 1–4: $FCF_{1-4} = 1,875,000$
* Year 5 -Terminal OCF plus NWC recovery: $FCF_5 = 1,875,000 + 300,000 = 2,175,000$

$$NPV = -6,300,000 + \sum_{t=1}^{4} \frac{1,875,000}{(1.10)^t} + \frac{2,175,000}{(1.10)^5}$$
$$NPV = -6,300,000 + 1,875,000 \times \left[ \frac{1 - (1.10)^{-4}}{0.10} \right] + \frac{2,175,000}{(1.10)^5} \quad \Rightarrow \quad NPV = 994,001.58$$

Since the NPV is positive $(\$994,001.58)$, Delta Semiconductor should proceed with the investment.

### Question 4.3

Omega Manufacturing produces 200,000 specialized automotive modules per year in its steady-state production facility. Currently, the company outsources the production of the internal logic controllers for these modules to a supplier at a price of $\$10.00$ each. The engineering team has proposed a plan to manufacture these controllers in-house, which would involve a variable production cost of $\$7.50$ per unit.

Calculate the Net Present Value (NPV) of the decision to move production in-house versus continuing to buy the components from the external supplier.

**Annual Pre-tax Cost Savings**
$$\text{Annual Savings} = (\text{Outsource Price} - \text{In-house Cost}) \times \text{Volume} \\ \quad \Rightarrow \quad \text{Annual Savings} = (\$10.00 - \$7.50) \times 200,000 = \$500,000$$

*   **Year 0:** $-\$800,000 (\text{Machinery}) - \$100,000 (\text{NWC}) = -\$900,000$

### Question 5.1

Sigma Robotics has 6.8 percent coupon bonds outstanding with 15 years remaining until maturity. The bonds make semiannual interest payments and are currently trading at 104 percent of their face value. 

Calculate the current yield, the yield to maturity (YTM), and the effective annual yield (EAY) for these bonds.

---

### Question 5.1: Solution

* Annual Coupon = 6.8% of $1,000 = $68.00
* Current Price = 104% of $1,000 = $1,040.00

$$\text{Current Yield} = \frac{\text{Annual Coupon}}{\text{Price}} \quad \Rightarrow \quad \text{Current Yield} = \frac{68}{1,040} \approx 0.065385 \text{ or } 6.54\%$$

$$1,040 = 34 \times \left[ \frac{1 - (1+r)^{-30}}{r} \right] + \frac{1,000}{(1+r)^{30}} \quad \Rightarrow \quad r \approx 0.03190856$$

The YTM (Bond-Equivalent Yield) is the periodic rate annualized as an APR:

$$\text{YTM} = 0.03190856 \times 2 \approx 0.063817 \text{ or } 6.3817\%$$

The EAY accounts for the compounding of the semiannual periods within a year:
$$\text{EAY} = (1 + r)^2 - 1 \quad \Rightarrow \quad \text{EAY} = (1.03190856)^2 - 1 \approx 0.064835 \text{ or } 6.4835\%$$

### Question 5.2

The Yield to Maturity (YTM) represents the total return anticipated on a bond if it is held until maturity. However, if market interest rates fluctuate and the bond is sold prior to maturity, the actual return achieved is known as the Holding Period Yield (HPY).

*   Industrial Power Corp recently issued 10-year bonds. You purchase one of these bonds today for $1,080. The bond pays an annual coupon of 7.5% and has a face value of $1,000. What is the initial YTM expected on this investment?

*   Suppose that exactly four years after your purchase, the prevailing YTM for similar bonds has risen by 1.00% from your initial purchase yield. You decide to sell your bond at this point. 
    
    *   What market price can you expect to receive for your bond at the end of Year 4? 
    
    *   Calculate the annualized HPY for your investment over the four-year holding period. 
    
    *   How does this HPY compare to the YTM calculated when you initially bought the bond? Discuss your results. 

---

### Question 5.2: Solution

**Initial YTM**
$$1,080 = 75 \times \left[ \frac{1 - (1 + r)^{-10}}{r} \right] + \frac{1,000}{(1 + r)^{10}} \quad \Rightarrow \quad \text{Initial YTM} \approx 6.3927\%$$


**Market Price at Year 4**

After four years, the bond has 6 years remaining until maturity ($10 - 4 = 6$). The new market yield is the initial YTM plus 1.00%:

$$\text{New YTM} = 6.3927\% + 1.00\% = 7.3927\%$$

$$P_4 = 75 \times \left[ \frac{1 - (1.073927)^{-6}}{0.073927} \right] + \frac{1,000}{(1.073927)^6} \quad \Rightarrow \quad  P_4 \approx 352.49 + 652.56 = \$1,005.05$$

**Annualized Holding Period Yield (HPY)**

The HPY is the internal rate of return for the 4-year holding period. You paid $1,080, received four annual coupons of $75, and sold the bond for $1,005.05.

$$1,080 = 75 \times \left[ \frac{1 - (1 + HPY)^{-4}}{HPY} \right] + \frac{1,005.05}{(1 + HPY)^4} \quad \Rightarrow \quad \text{Annualized HPY} \approx 5.3425\%$$

Theta Innovations has issued zero-coupon corporate bonds with an 8-year maturity and a face value of $1,000.

### Question 5.5

Sigma Global Logistics operates a specialized supply chain network across Central Europe. It has just issued a four-year, zero-coupon corporate bond at a market price of $84. You have purchased this bond and plan to hold it until it matures. Assume the face value is $100.

a) Calculate the yield to maturity (YTM) of this bond.

b) What is the expected annual return on your investment if there is zero probability of default?

c) What is your expected annual return if default is certain (100% probability), and you anticipate recovering 75% of the face value at the end of four years?

d) What is your expected annual return if the probability of default occurring before maturity is 35%, and if default occurs, you will recover 75% of the face value?

e) Considering your answers for parts (b) through (d), comment on the relationship between the expected return and the yield to maturity in the presence of default risk.

---

### Question 5.5: Solution

$$84 = \frac{100}{(1 + YTM)^4} \quad \Rightarrow \quad (1 + YTM)^4 = \frac{100}{84} \approx 1.19047619 \quad \Rightarrow \quad YTM \approx 4.4552\%$$

If the probability of default is 0%, the expected cash flow equals the promised cash flow ($100).
$$Expected \ Return = YTM \approx 4.4552\%$$

**Expected annual return if default is certain (100% probability)**
If default is certain and recovery is 75%:
$$84 = \frac{75}{(1 + r)^4} \quad \Rightarrow \quad (1 + r)^4 = \frac{75}{84} \approx 0.89285714 \quad \Rightarrow \quad r \approx -2.7935\%$$

**Expected annual return if probability of default is 35%**
* 65% probability of receiving $100 (No default)
* 35% probability of receiving $75 (Default)
$$E[CF] = (0.65 \times \$100) + (0.35 \times \$75) = \$65 + \$26.25 = \$91.25$$

$$84 = \frac{91.25}{(1 + r)^4} \quad \Rightarrow \quad (1 + r)^4 = \frac{91.25}{84} \approx 1.08630952 \quad \Rightarrow \quad r \approx 2.0912\%$$

As the risk of default increases, the expected return on the bond decreases, falling further below the YTM. 
* The YTM is a "promised" return that assumes the issuer will definitely pay the full face value. 
* However, the market price ($84) reflects the market's assessment of risk. 
* Because the expected cash flows ($91.25) are lower than the promised cash flows ($100), the YTM will always overstate the actual expected return when default risk is present.
